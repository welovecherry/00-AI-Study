```python
import pandas as pd

# 첫 번째 데이터프레임: 학생들의 이름과 시험 점수
df1 = pd.DataFrame({
    '이름': ['김철수', '박영희', '이지수', '최민준'],
    '시험점수': [85, 92, 78, 96]
})

# 두 번째 데이터프레임: 학생들의 이름과 학급 정보
df2 = pd.DataFrame({
    '이름': ['김철수', '박영희', '이지수', '정다은'],
    '학급': ['1반', '2반', '1반', '3반']
})

print("첫 번째 데이터프레임 (df1):")
print(df1)
print("\n두 번째 데이터프레임 (df2):")
print(df2)
```

```python
result = df1.merge(df2, how='left', on='이름')
result
```

```python
# how='left': 왼쪽 데이터프레임(df1)의 모든 행을 유지 (VLOOKUP과 유사)
inner_res = df1.merge(df2, how='inner', on='이름')
inner_res
```

```python
outer_res = df1.merge(df2, how='outer', on='이름')
outer_res
```

```python
df1 = pd.DataFrame({
    '이름': ['김철수', '김철수', '박영희', '박영희'],
    '학년': [1, 2, 1, 2],
    '시험점수': [85, 92, 78, 96]
})

# 두 번째 데이터프레임: 학생들의 이름, 학년, 학급 정보
df2 = pd.DataFrame({
    '이름': ['김철수', '김철수', '박영희', '이지수'],
    '학년': [1, 2, 1, 1],
    '학급': ['A반', 'B반', 'A반', 'C반']
})
print(df1)
print(df2)

res = df1.merge(df2, how='left', on=['이름', '학년'])
res
```
