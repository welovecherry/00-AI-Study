 ## 📚 핵심 자료구조 복습 (List, Dict, Set) 🧠🔄



 알고리즘 문제를 효율적으로 풀기 위해서는 문제 상황에 맞는 적절한 '도구', 즉 자료구조를 선택하는 것이 매우 중요합니다. 어제 배운 핵심 자료구조들을 빠르게 복습해 봅시다.



 | 자료구조          | 주요 특징                                     | 강점 (효율적인 연산)                     | 약점 (비효율적인 연산)                   |
 | :---------------- | :-------------------------------------------- | :--------------------------------------- | :--------------------------------------- |
 | **List `[]`** | 순서 O, 변경 O, 중복 O                        | **인덱스 접근 `[i]` (O(1))**, 순차 순회      | **값 탐색 `in` (O(n))**, 중간 삽입/삭제 (O(n)) |
 | **Dict `{}`** | **키-값 쌍**, 키 고유, (3.7+ 순서 O)           | **키 기반 검색/삽입/삭제 (평균 O(1))** ⚡️   | 키 외의 값 검색 (O(n)), 메모리 사용량↑ |
 | **Set `{}`** | 순서 X, 변경 O, **중복 X** | **멤버십 테스트 `in` (평균 O(1))**⚡️, 중복 제거 | 순서 필요 작업 불가, 요소 접근 불가    |



 **🤔 언제 무엇을 쓸까?**



 * **순서가 중요**하고, **인덱스로 빠르게 접근**해야 할 때 → **List**

 * **고유한 키**를 통해 **매우 빠르게 값**을 찾거나 저장/삭제해야 할 때 → **Dict**

 * **중복을 제거**해야 하거나, 어떤 요소가 **존재하는지 매우 빠르게 확인**해야 할 때 → **Set**

 ## ✨ 복잡도의 중요성 재확인 ⏱️💾



 * **시간 복잡도 (Time Complexity)**: 입력 데이터 크기(n) 증가에 따른 **실행 시간** 증가율. (얼마나 빠른가?)

 * **공간 복잡도 (Space Complexity)**: 입력 데이터 크기(n) 증가에 따른 **메모리 사용량** 증가율. (메모리 얼마나 쓰는가?)



 * **왜 중요할까?**

     * 동일한 문제라도 알고리즘(그리고 사용된 자료구조)에 따라 성능 차이가 **매우 크게** 발생! ⚡️ vs 🐢

     * 데이터 크기가 커질수록 복잡도가 낮은 알고리즘의 중요성은 **기하급수적으로 증가**함.

     * **O(1)** < **O(log n)** < **O(n)** < **O(n log n)** < **O(n²)** < **O(2ⁿ)** ... (왼쪽일수록 좋음)

     * 특히 **코딩 테스트**나 **대규모 데이터 처리** 시 복잡도 고려는 필수!



 * **자료구조 선택과 복잡도**: 어떤 자료구조를 선택하느냐가 알고리즘의 전체 복잡도를 결정하는 핵심 요인 중 하나!

     * 예: 리스트에서 값 찾기 O(n) vs 딕셔너리/셋에서 키/값 찾기 O(1)

 ### 문제 1: 할 일 목록 관리하기 (리스트 메서드) (난이도: ⭐️)



 **상황:** 파이썬 `list` 로 오늘의 할 일 목록(`tasks`)을 관리하고 있습니다.



 **요구사항:**

 1.  목록 맨 뒤에 '저녁 식사 준비'를 추가하세요.

 2.  목록에서 2번 인덱스(세 번째 항목)의 할 일을 제거하고, 제거된 항목을 확인하세요.

 3.  '이메일 확인' 항목이 몇 번 인덱스에 있는지 찾아보세요.



 **데이터 예시:**

 ```python

 tasks = ['파이썬 공부', '이메일 확인', '운동하기']

 ```



 ### 🤔 어떤 메서드를 사용하는 것이 좋을까요?

 * 리스트 끝에 항목 추가하기

 * 특정 위치의 항목 제거하고 값 얻기

 * 특정 값의 위치(인덱스) 찾기

 ➡️ 리스트의 `.append()`, `.pop()`, `.index()` 메서드가 필요해 보입니다!

 ### 힌트

 * 맨 뒤에 추가: `리스트이름.append(추가할_항목)`

 * 특정 인덱스 제거 및 확인: `제거된_항목 = 리스트이름.pop(인덱스번호)`

 * 값으로 인덱스 찾기: `인덱스 = 리스트이름.index(찾을_값)`

 ### 코드 조각 (메서드 활용)

````python
# %%
tasks = ['파이썬 공부', '이메일 확인', '운동하기']
print(f"초기 할 일 목록: {tasks}")

# 1. '저녁 식사 준비' 추가
tasks.append('저녁 식사 준비')
print(f"1. 추가 후: {tasks}")

# 2. 2번 인덱스 항목 제거 및 확인
removed_task = tasks.pop(2) # '운동하기' 제거됨
print(f"2. 제거된 항목: '{removed_task}', 제거 후 목록: {tasks}")

# 3. '이메일 확인' 인덱스 찾기
try: # .index()는 찾는 값이 없으면 오류 발생하므로 try-except 사용 가능
    email_index = tasks.index('이메일 확인')
    print(f"3. '이메일 확인'의 인덱스: {email_index}")
except ValueError:
    print("3. '이메일 확인' 항목이 목록에 없습니다.")
````

 ### 문제 2: 중복 없는 태그 관리 (세트 메서드) (난이도: ⭐️)



 **상황:** 게시글에 사용된 태그(`tags`)들을 파이썬 `set`으로 관리하여 중복을 없애고, 특정 태그 존재 여부를 확인하려 합니다.



 **요구사항:**

 1.  태그 모음(`tags`)에 '#파이썬' 태그를 추가하세요 (이미 있다면 변화 없음).

 2.  '#알고리즘' 태그가 이미 `tags`에 있는지 확인하세요.

 3.  두 개의 태그 세트 `tags1`과 `tags2`의 차이(tags1에는 있고 tags2에는 없는 태그)를 구하세요.



 **데이터 예시:**

 ```python

 tags = {'#AI', '#데이터분석', '#파이썬'}

 tags1 = {'#AI', '#파이썬', '#머신러닝'}

 tags2 = {'#파이썬', '#데이터분석', '#딥러닝'}

 ```



 ### 🤔 어떤 메서드/연산자를 사용하는 것이 좋을까요?

 * 세트에 원소 추가하기 (중복은 자동 처리)

 * 세트에 특정 원소가 있는지 빠르게 확인하기

 * 한 세트에는 있고 다른 세트에는 없는 원소들 찾기

 ➡️ 세트의 `.add()` 메서드, `in` 연산자, `-` 연산자(또는 `.difference()` 메서드)가 유용해 보입니다!

 ### 힌트

 * 원소 추가: `세트이름.add(추가할_원소)`

 * 원소 존재 확인: `찾을_원소 in 세트이름` (결과는 True 또는 False)

 * 차집합: `세트1 - 세트2` 또는 `세트1.difference(세트2)`

 ### 코드 조각 (메서드/연산자 활용)

````python
# %%
tags = {'#AI', '#데이터분석', '#파이썬'}
print(f"초기 태그: {tags}")

# 1. '#파이썬' 태그 추가 시도
tags.add('#파이썬') # 이미 있으므로 변화 없음
tags.add('#알고리즘') # 새로운 태그 추가
print(f"1. 태그 추가 후: {tags}")

# 2. '#알고리즘' 태그 존재 확인
has_algo_tag = '#알고리즘' in tags
print(f"2. '#알고리즘' 태그가 있는가? {has_algo_tag}")

# 3. 두 세트의 차집합 구하기
tags1 = {'#AI', '#파이썬', '#머신러닝'}
tags2 = {'#파이썬', '#데이터분석', '#딥러닝'}
difference_tags = tags1 - tags2 # tags1 에만 있는 태그
print(f"3. tags1 - tags2 차집합: {difference_tags}")
````

 ### 문제 3: 학생 점수 관리 (딕셔너리 기본 연산) (난이도: ⭐️)



 **상황:** 학생들의 시험 점수를 파이썬 `dict` 로 관리합니다 (`scores`).



 **요구사항:**

 1.  'Bob'의 점수를 확인하세요.

 2.  'David'의 점수를 95점으로 추가(또는 수정)하세요.

 3.  'Charlie'라는 학생이 점수 목록에 있는지 확인하세요.



 **데이터 예시:**

 ```python

 scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78}

 ```



 ### 🤔 어떤 방법을 사용하는 것이 좋을까요?

 * 키(학생 이름)를 이용해 값(점수)에 접근하기

 * 새로운 키-값 쌍을 추가하거나, 기존 키의 값 수정하기

 * 딕셔너리에 특정 키가 존재하는지 확인하기

 ➡️ 딕셔너리의 `[key]` 접근, `[key] = value` 할당, `key in dictionary` 연산자를 사용하면 됩니다!

 ### 힌트

 * 값 접근: `딕셔너리이름[키]` (키가 없으면 오류 발생!) 또는 `딕셔너리이름.get(키, 기본값)`

 * 추가/수정: `딕셔너리이름[키] = 새로운값`

 * 키 존재 확인: `키 in 딕셔너리이름` (결과는 True 또는 False)

 ### 코드 조각 (딕셔너리 활용)

````python
# %%
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78}
print(f"초기 점수: {scores}")

# 1. 'Bob'의 점수 확인
bob_score = scores.get('Bob', 'N/A') # .get() 사용 시 키 없어도 오류 안 남
print(f"1. Bob의 점수: {bob_score}")

# 2. 'David'의 점수 추가/수정
scores['David'] = 95
print(f"2. David 점수 추가 후: {scores}")
scores['Alice'] = 88 # 기존 값 수정
print(f"   Alice 점수 수정 후: {scores}")

# 3. 'Charlie' 존재 확인
has_charlie = 'Charlie' in scores
print(f"3. Charlie가 목록에 있는가? {has_charlie}")
has_eve = 'Eve' in scores
print(f"   Eve가 목록에 있는가? {has_eve}")
````

 ### 문제 4: 과일 판매량 집계 (Counter 사용) (난이도: ⭐️⭐️)



 **상황:** 오늘 판매된 과일 목록(`sold_fruits`)이 리스트로 있습니다. 각 과일이 몇 개씩 팔렸는지 쉽게 집계하고, 가장 많이 팔린 과일을 알고 싶습니다.



 **요구사항:**

 1.  `sold_fruits` 리스트를 이용하여 각 과일의 판매량(빈도수)을 계산하세요.

 2.  가장 많이 팔린 과일과 그 개수를 찾으세요.



 **데이터 예시:**

 ```python

 sold_fruits = ['사과', '바나나', '사과', '오렌지', '바나나', '사과', '포도', '사과']

 ```



 ### 🤔 어떤 자료구조와 메서드를 사용하는 것이 좋을까요?

 * 리스트 내 항목들의 개수를 세는 데 특화된 자료구조가 필요해요.

 * 가장 빈도가 높은 항목을 쉽게 찾는 기능이 있으면 좋겠어요.

 ➡️ `collections.Counter` 객체를 생성하고, `.most_common()` 메서드를 활용하면 편리합니다!

 ### 힌트

 * `from collections import Counter`

 * 빈도수 계산: `카운터객체 = Counter(리스트)`

 * 가장 많이 팔린 항목 (상위 1개): `카운터객체.most_common(1)` (결과는 `[(항목, 개수)]` 형태의 리스트)

 ### 코드 조각 (Counter 활용)

````python
# %%
from collections import Counter

sold_fruits = ['사과', '바나나', '사과', '오렌지', '바나나', '사과', '포도', '사과']
print(f"판매된 과일 목록: {sold_fruits}")

# 1. 각 과일 판매량 집계
fruit_counts = Counter(sold_fruits)
print(f"1. 과일별 판매량: {fruit_counts}")
# 예: Counter({'사과': 4, '바나나': 2, '오렌지': 1, '포도': 1})

# 2. 가장 많이 팔린 과일 찾기
most_common_fruit = fruit_counts.most_common(1) # 가장 빈번한 1개
if most_common_fruit: # 결과가 있을 경우 (빈 리스트가 아닐 경우)
    top_fruit, top_count = most_common_fruit[0] # 튜플 언패킹
    print(f"2. 가장 많이 팔린 과일: '{top_fruit}', 개수: {top_count}개")
else:
    print("2. 판매된 과일이 없습니다.")
````

 ### 문제 5: 카테고리별 상품 목록 만들기 (defaultdict 사용) (난이도: ⭐️⭐️)



 **상황:** 여러 상품 정보(`products`)가 `(카테고리, 상품명)` 튜플의 리스트로 주어졌습니다. 이를 카테고리별로 묶어서 딕셔너리로 만들고 싶습니다.



 **요구사항:**

 `products` 리스트를 순회하며, `defaultdict(list)`를 사용하여 각 카테고리(key)에 해당하는 상품명들을 리스트(value)로 추가하세요.



 **데이터 예시:**

 ```python

 products = [('과일', '사과'), ('음료', '콜라'), ('과일', '바나나'), ('과자', '새우깡'), ('음료', '사이다'), ('과일', '포도')]

 ```

 **예상 결과 (딕셔너리):**

 ```

 {'과일': ['사과', '바나나', '포도'], '음료': ['콜라', '사이다'], '과자': ['새우깡']}

 ```



 ### 🤔 어떤 자료구조와 메서드를 사용하는 것이 좋을까요?

 * 결과를 저장할 구조는 키(카테고리)와 값(상품 리스트)을 가져야 해요. -> 딕셔너리!

 * 딕셔너리에 키가 없을 때마다 빈 리스트를 만들어줘야 하는 번거로움을 피하고 싶어요.

 ➡️ `collections.defaultdict(list)` 를 사용하면 키가 없을 때 자동으로 빈 리스트를 생성해주므로 코드가 간결해집니다! 리스트에 항목을 추가할 때는 `.append()` 메서드를 사용합니다.

 ### 힌트

 * `from collections import defaultdict`

 * 딕셔너리 생성: `카테고리별_상품 = defaultdict(list)`

 * `products` 리스트를 `for` 루프로 순회하며 `category, product_name`을 얻습니다.

 * `카테고리별_상품[category].append(product_name)` 코드로 해당 카테고리 리스트에 상품명을 추가합니다. (key가 없어도 `defaultdict`가 자동으로 빈 리스트를 만들어줍니다.)

 ### 코드 조각 (defaultdict 활용)

````python
# %%
from collections import defaultdict

products = [('과일', '사과'), ('음료', '콜라'), ('과일', '바나나'), ('과자', '새우깡'), ('음료', '사이다'), ('과일', '포도')]
print(f"상품 목록: {products}")

# defaultdict(list) 생성
products_by_category = defaultdict(list)

# products 리스트를 순회하며 카테고리별로 상품 추가
for category, product_name in products:
    products_by_category[category].append(product_name)

print("\n카테고리별 상품 목록 (defaultdict):")
print(products_by_category)
# 보기 좋게 일반 dict로 변환 출력
print("\n카테고리별 상품 목록 (dict 변환):")
print(dict(products_by_category))
````

 ### 문제 6: 가장 작은 작업 먼저 처리하기 (heapq 사용) (난이도: ⭐️⭐️⭐️)



 **상황:** 처리해야 할 작업들이 `(우선순위, 작업명)` 튜플 형태로 리스트 `tasks`에 저장되어 있습니다. 우선순위 숫자가 **작을수록** 먼저 처리해야 합니다. 가장 우선순위가 높은 작업을 효율적으로 찾고 제거하고 싶습니다.



 **요구사항:**

 1.  `tasks` 리스트를 최소 힙(min-heap) 구조로 만듭니다. (파이썬 `heapq`는 리스트를 사용)

 2.  힙에서 가장 우선순위가 높은(숫자가 가장 작은) 작업을 꺼내서 확인합니다.

 3.  새로운 작업 `(2, '보고서 작성')` 을 힙에 추가합니다.

 4.  다시 가장 우선순위가 높은 작업을 꺼내서 확인합니다.



 **데이터 예시:**

 ```python

 tasks = [(5, '이메일 발송'), (1, '긴급 버그 수정'), (3, '데이터 백업')]

 ```



 ### 🤔 어떤 모듈과 함수를 사용하는 것이 좋을까요?

 * 여러 항목 중 가장 작은 값을 빠르게 찾고 제거/추가하는 데 효율적인 구조가 필요해요.

 * 파이썬에서 힙(Heap) 기능을 제공하는 표준 라이브러리 모듈이 있어요.

 ➡️ `heapq` 모듈을 사용해야 합니다! 리스트를 힙으로 사용하며, `heapq.heappush()`로 원소를 추가하고 `heapq.heappop()`으로 가장 작은 원소를 꺼냅니다. (리스트를 힙으로 변환하는 `heapq.heapify()`도 유용합니다.)

 ### 힌트

 * `import heapq`

 * 리스트를 힙으로 변환 (선택적이지만 효율적): `heapq.heapify(리스트이름)`

 * 힙에 원소 추가: `heapq.heappush(힙리스트, 추가할원소)` (튜플 넣으면 첫 요소 기준)

 * 힙에서 가장 작은 원소 꺼내기: `가장작은원소 = heapq.heappop(힙리스트)`

 * (주의: `heapq`는 최소 힙이므로, 숫자가 작을수록 우선순위가 높다는 문제 상황과 잘 맞습니다.)

 ### 코드 조각 (heapq 활용)

````python
# %%
import heapq

tasks = [(5, '이메일 발송'), (1, '긴급 버그 수정'), (3, '데이터 백업')]
print(f"초기 작업 목록 (리스트): {tasks}")

# 1. 리스트를 최소 힙으로 변환 (튜플의 첫 번째 요소인 우선순위 기준)
heapq.heapify(tasks)
print(f"힙으로 변환된 리스트: {tasks}") # 힙 구조로 내부 순서 변경됨

# 2. 가장 우선순위 높은 작업 꺼내기
if tasks: # 힙이 비어있지 않다면
    highest_priority_task = heapq.heappop(tasks)
    print(f"2. 가장 우선순위 높은 작업: {highest_priority_task}")
    print(f"   작업 꺼낸 후 힙: {tasks}")
else:
    print("2. 처리할 작업이 없습니다.")


# 3. 새로운 작업 추가
new_task = (2, '보고서 작성')
heapq.heappush(tasks, new_task)
print(f"3. '{new_task}' 추가 후 힙: {tasks}")

# 4. 다시 가장 우선순위 높은 작업 꺼내기
if tasks:
    next_highest_priority_task = heapq.heappop(tasks)
    print(f"4. 다음 우선순위 높은 작업: {next_highest_priority_task}")
    print(f"   작업 꺼낸 후 힙: {tasks}")
else:
    print("4. 처리할 작업이 없습니다.")
````

 ### 문제 1: 참가자 명단 정리 (난이도: ⭐️⭐️)



 **상황:** 이벤트에 참가 신청한 사람들의 명단(`participants`)이 있습니다. 그런데 중복 신청한 사람이나 신청 철회한 사람(`blacklist`) 명단이 섞여 있을 수 있습니다. 최종적으로 이벤트에 참여할 수 있는 **고유한 참가자**들의 명단(알파벳 순 정렬)을 만들어야 합니다.



 **요구사항:**

 1.  `participants` 리스트와 `blacklist` 리스트를 입력받습니다.

 2.  `blacklist`에 포함된 사람은 제외합니다.

 3.  중복된 이름은 한 번만 포함합니다.

 4.  최종 참가자 명단을 알파벳 순서로 정렬하여 반환합니다.



 **데이터 예시:**

 * `participants = ["Leo", "Kai", "Leo", "Emma", "Chloe"]`

 * `blacklist = ["Kai", "Chloe"]`

 * → `["Emma", "Leo"]` (Leo 중복 제거, Kai/Chloe 블랙리스트 제거, Emma 남음, 알파벳 순 정렬)



 * `participants = ["John", "Jane", "John", "Peter"]`

 * `blacklist = ["Jane"]`

 * → `["John", "Peter"]`



 ### 🤔 어떤 자료구조를 사용하는 것이 좋을까요?

 * **빠르게 특정 참가자가 블랙리스트에 있는지 확인**해야 합니다. (멤버십 테스트)

 * **중복된 이름을 효율적으로 제거**해야 합니다.

 ➡️ **집합(Set)** 자료구조가 '포함 여부 확인'과 '중복 제거'에 매우 효율적입니다! 블랙리스트를 집합으로 만들고, 참가자 명단도 집합으로 만들어 차집합 연산을 활용하면 어떨까요? 마지막 정렬은 리스트로 변환 후 수행하면 됩니다.



 **비교:**

 * **리스트만 사용:** 각 참가자에 대해 블랙리스트 전체를 순회하며 확인 (O(n*m)), 중복 제거를 위해 또 다른 리스트나 로직 필요 (비효율적).

 * **집합 사용:** 블랙리스트를 집합으로 변환 (O(m)). 참가자 명단을 집합으로 변환 (O(n)). 차집합 연산 (평균 O(n+m)). 최종 정렬 O(k log k). 전체적으로 훨씬 효율적.

 ### 힌트

 * `blacklist` 리스트를 `set()`을 이용하여 집합 `blacklist_set`으로 변환합니다. 이렇게 하면 특정 인물이 블랙리스트에 있는지 O(1) 평균 시간에 확인할 수 있습니다.

 * `participants` 리스트도 `set()`을 이용하여 집합 `participants_set`으로 변환합니다. 이 과정에서 자동으로 중복된 이름이 제거됩니다.

 * 두 집합의 **차집합** 연산 (`participants_set - blacklist_set`)을 사용하여 `participants_set`에서 `blacklist_set`에 포함된 요소를 제거합니다. 결과 역시 집합입니다.

 * 최종 결과를 `list()`를 이용해 리스트로 변환하고, `sorted()` 함수 또는 `.sort()` 메서드를 사용하여 알파벳 순으로 정렬합니다.

 ### 정답 코드

````python
# %%
import collections # collections.deque를 사용하기 위해 import (다른 문제에서 필요)

def finalize_participants(participants: list[str], blacklist: list[str]) -> list[str]:
    """블랙리스트 제외, 중복 제거 후 정렬된 최종 참가자 명단 반환 (집합 활용)"""

    # 집합 연산을 활용한 간결한 버전
    participants_set = set(participants)
    blacklist_set = set(blacklist)

    # 차집합 연산: 참가자 중 블랙리스트에 있는 사람 제외 (중복도 자동 처리됨)
    final_set = participants_set - blacklist_set

    # 결과를 리스트로 변환 후 정렬하여 반환
    return sorted(list(final_set))

# 테스트 케이스
participants1 = ["Leo", "Kai", "Leo", "Emma", "Chloe"]
blacklist1 = ["Kai", "Chloe"]
print(f"참가자: {participants1}, 블랙리스트: {blacklist1} -> 최종 명단: {finalize_participants(participants1, blacklist1)}")

participants2 = ["John", "Jane", "John", "Peter"]
blacklist2 = ["Jane"]
print(f"참가자: {participants2}, 블랙리스트: {blacklist2} -> 최종 명단: {finalize_participants(participants2, blacklist2)}")
````

 ### 문제 2: 최근 N개 작업 기록 유지 (난이도: ⭐️⭐️)



 **상황:** 사용자의 작업 기록을 시스템에 저장하려고 합니다. 하지만 모든 기록을 저장하기에는 용량이 부족하여, **가장 최근 N개의 작업 기록만 순서대로 유지**하고 싶습니다. 새로운 작업이 추가되면 가장 오래된 작업 기록은 삭제되어야 합니다.



 **요구사항:**

 1.  유지할 최대 작업 기록 개수 `N`과 사용자의 작업 순서(`actions`) 리스트를 입력받습니다.

 2.  `actions` 리스트의 작업을 순서대로 처리하며, 항상 최대 `N`개의 최근 작업만 순서대로 유지하는 리스트를 반환합니다.



 **데이터 예시:**

 * `N = 3`, `actions = ["login", "edit", "save", "logout", "send_email"]`

 * 처리 과정:

     * `["login"]`

     * `["login", "edit"]`

     * `["login", "edit", "save"]`

     * `["edit", "save", "logout"]` (login 삭제)

     * `["save", "logout", "send_email"]` (edit 삭제)

 * → 최종 결과: `["save", "logout", "send_email"]`



 ### 🤔 어떤 자료구조를 사용하는 것이 좋을까요?

 * 작업은 **순서대로** 기록되어야 합니다.

 * 새로운 작업은 **오른쪽(끝)** 에 추가됩니다.

 * 기록 개수가 `N`개를 초과하면 **왼쪽(처음)** 에서 가장 오래된 기록을 제거해야 합니다.

 ➡️ 양쪽 끝에서의 추가/삭제가 빈번하게 일어납니다! **데크(Deque)** 가 양방향 O(1) 추가/삭제를 지원하므로 가장 효율적입니다. `collections.deque`를 사용할 때 `maxlen` 옵션을 주면 크기 제한과 오래된 요소 자동 제거를 쉽게 구현할 수 있습니다.



 **비교:**

 * **리스트 사용:** 오른쪽 끝 추가(`append`)는 O(1)이지만, 왼쪽 처음 요소 삭제(`pop(0)`)는 O(k) (k는 리스트 길이)여서 작업이 많아지면 비효율적입니다.

 * **데크 사용:** 오른쪽 끝 추가(`append`)와 왼쪽 처음 요소 삭제(`popleft`) 모두 O(1)이므로 매우 효율적입니다. `maxlen` 사용 시 크기 관리도 자동으로 됩니다.

 ### 힌트

 * `from collections import deque`를 사용하여 데크 자료구조를 가져옵니다.

 * `deque(maxlen=N)` 와 같이 `maxlen` 인자를 주어 데크를 생성합니다. 이렇게 하면 데크의 최대 크기가 `N`으로 고정됩니다.

 * 주어진 `actions` 리스트를 순회하면서 각 `action`을 `deque.append(action)` 메서드를 사용하여 데크의 오른쪽에 추가합니다.

 * 만약 `append` 시 데크의 크기가 `maxlen`을 초과하게 되면, 데크는 자동으로 **왼쪽 끝(가장 오래된) 요소**를 제거합니다.

 * 모든 `actions`를 처리한 후, 최종 상태의 데크를 `list()`를 이용하여 리스트로 변환하여 반환합니다.

 ### 정답 코드

````python
# %%
from collections import deque # 이미 위에서 import 했지만, 명확성을 위해 다시 표기

def get_recent_actions(n: int, actions: list[str]) -> list[str]:
    """가장 최근 N개의 작업 기록만 순서대로 유지하여 반환 (데크 활용)"""

    # 최대 크기 N을 가지는 데크 생성 (maxlen 옵션 활용)
    # maxlen이 지정된 deque는 크기가 초과되면 자동으로 반대편 끝의 요소를 제거함
    recent_actions_deque = deque(maxlen=n)

    # 작업 기록을 순서대로 데크에 추가
    for action in actions:
        recent_actions_deque.append(action) # 오른쪽에 추가, N개 넘으면 왼쪽 자동 제거

    # 최종적으로 데크에 남아있는 요소들을 리스트로 변환하여 반환
    return list(recent_actions_deque)

# 테스트 케이스
n1 = 3
actions1 = ["login", "edit", "save", "logout", "send_email"]
print(f"N={n1}, Actions={actions1} -> 최근 기록: {get_recent_actions(n1, actions1)}")

n2 = 2
actions2 = ["A", "B", "C", "D"]
print(f"N={n2}, Actions={actions2} -> 최근 기록: {get_recent_actions(n2, actions2)}")
````

 ### 문제 3: 상품 재고 관리 (난이도: ⭐️⭐️⭐️)



 **상황:** 온라인 쇼핑몰의 상품 재고를 관리하는 시스템을 만들고 있습니다. 각 상품(ID)마다 현재 남아있는 재고 수량을 추적해야 합니다. 입고(`stock_in`) 시 재고가 증가하고, 주문(`order`) 시 재고가 감소합니다. 특정 상품의 현재 재고를 빠르게 조회할 수 있어야 합니다.



 **요구사항:**

 1.  초기 재고 상태(`initial_stock`) 딕셔너리와 변경 기록(`updates`) 리스트를 입력받습니다.

 2.  `updates`는 `("상품ID", "행동", 수량)` 튜플의 리스트입니다. ("행동"은 "stock_in" 또는 "order")

 3.  모든 변경 기록을 처리한 후, 특정 상품 ID(`query_item`)의 최종 재고 수량을 반환합니다. 만약 해당 상품 ID가 없다면 0을 반환합니다.



 **데이터 예시:**

 * `initial_stock = {"item1": 10, "item2": 5}`

 * `updates = [("item1", "order", 3), ("item3", "stock_in", 7), ("item2", "order", 5), ("item1", "stock_in", 5)]`

 * `query_item = "item1"`

 * 처리 과정:

     * 초기: `{"item1": 10, "item2": 5}`

     * item1 주문 3: `{"item1": 7, "item2": 5}`

     * item3 입고 7: `{"item1": 7, "item2": 5, "item3": 7}` (새 상품 추가)

     * item2 주문 5: `{"item1": 7, "item2": 0, "item3": 7}`

     * item1 입고 5: `{"item1": 12, "item2": 0, "item3": 7}`

 * → `query_item` "item1"의 최종 재고: `12`



 * `query_item = "item4"` → `0` (존재하지 않으므로)



 ### 🤔 어떤 자료구조를 사용하는 것이 좋을까요?

 * **상품 ID를 이용해 해당 상품의 재고 수량에 빠르게 접근**하고 수정해야 합니다. (Key-Value 조회)

 * 새로운 상품(새로운 Key)이 추가될 수도 있습니다.

 ➡️ **딕셔너리(Dict)** 가 {상품 ID: 재고 수량} 형태로 데이터를 저장하고, 상품 ID(키)를 통해 재고(값)를 O(1) 평균 시간에 조회/수정하는 데 최적입니다. `get()` 메서드를 사용하면 키가 없을 때의 처리도 간편합니다.



 **비교:**

 * **리스트 사용:** `[("item1", 10), ("item2", 5)]` 와 같이 리스트에 튜플로 저장한다면, 특정 상품을 찾기 위해 리스트 전체를 순회해야 하므로(O(n)) 매우 비효율적입니다. 수정이나 새 상품 추가도 번거롭습니다.

 * **딕셔너리 사용:** `{상품ID: 재고}` 형태로 저장하면, `stock["item1"]` 또는 `stock.get("item1", 0)` 처럼 O(1) 평균 시간에 바로 재고에 접근하고 수정할 수 있습니다. 새 상품 추가도 `stock["new_item"] = quantity` 로 간단합니다.

 ### 힌트

 * `initial_stock` 딕셔너리를 복사하여 `current_stock` 딕셔너리를 만듭니다. (원본 수정을 피하기 위해) `.copy()` 메서드를 사용하세요.

 * `updates` 리스트를 순회하면서 각 `(item_id, action, quantity)` 튜플을 처리합니다.

 * `action`이 "stock_in" 이면, `current_stock`에서 `item_id`에 해당하는 값을 찾아 `quantity`만큼 더해줍니다. 만약 `item_id`가 `current_stock`에 없다면, 초기값을 0으로 간주하고 더해야 합니다. `current_stock[item_id] = current_stock.get(item_id, 0) + quantity` 와 같이 `get()` 메서드를 사용하면 편리합니다. (`get(key, default_value)`는 key가 없을 때 default_value를 반환)

 * `action`이 "order" 이면, 마찬가지로 `current_stock.get(item_id, 0)`를 이용해 현재 재고를 가져온 뒤 `quantity`만큼 빼줍니다.

 * 모든 `updates` 처리가 끝나면, `query_item`에 해당하는 최종 재고를 `current_stock.get(query_item, 0)`을 사용하여 조회하고 반환합니다. (조회하려는 상품이 없으면 0 반환)

 ### 정답 코드

````python
# %%
def manage_stock(initial_stock: dict[str, int], updates: list[tuple[str, str, int]], query_item: str) -> int:
    """상품 재고 변경사항을 반영하고 특정 상품의 최종 재고를 반환 (딕셔너리 활용)"""

    # 초기 재고를 복사하여 사용 (원본 딕셔너리 변경 방지)
    current_stock = initial_stock.copy()

    # 변경 기록을 순회하며 재고 업데이트
    for item_id, action, quantity in updates:
        # .get(key, 0) : key가 있으면 해당 값, 없으면 0을 반환하여 계산
        current_quantity = current_stock.get(item_id, 0)

        if action == "stock_in":
            current_stock[item_id] = current_quantity + quantity
        elif action == "order":
            current_stock[item_id] = current_quantity - quantity
            # 실제 시스템에서는 재고가 음수가 되지 않도록 추가 처리가 필요할 수 있음
            # 예: current_stock[item_id] = max(0, current_quantity - quantity)

    # 조회하려는 상품의 최종 재고 반환 (없으면 0)
    return current_stock.get(query_item, 0)

# 테스트 케이스
initial_stock1 = {"item1": 10, "item2": 5}
updates1 = [("item1", "order", 3), ("item3", "stock_in", 7), ("item2", "order", 5), ("item1", "stock_in", 5)]
query_item1 = "item1"
print(f"초기 재고: {initial_stock1}, 변경: {updates1}")
print(f"-> {query_item1}의 최종 재고: {manage_stock(initial_stock1, updates1, query_item1)}")

query_item2 = "item3"
print(f"-> {query_item2}의 최종 재고: {manage_stock(initial_stock1, updates1, query_item2)}")

query_item3 = "item4"
print(f"-> {query_item3}의 최종 재고: {manage_stock(initial_stock1, updates1, query_item3)}")
````

 ### 문제 4: 긴급 요청 처리 (난이도: ⭐️⭐️⭐️)



 **상황:** 여러 부서에서 시스템 처리 요청이 들어옵니다. 각 요청은 `(우선순위, 요청 내용)` 형태로 주어지며, **우선순위 숫자가 낮을수록 더 긴급**합니다. 시스템은 항상 가장 긴급한 요청(가장 낮은 우선순위 숫자)부터 처리해야 합니다.



 **요구사항:**

 1.  여러 개의 요청 `requests` 리스트 `[(우선순위, 요청 내용), ...]` 를 입력받습니다.

 2.  모든 요청을 받아서, 처리해야 할 순서대로(우선순위가 낮은 순서, 같다면 순서는 무관) 요청 내용을 리스트로 반환합니다.



 **데이터 예시:**

 * `requests = [(3, "보고서 생성"), (1, "서버 점검"), (2, "데이터 백업"), (1, "긴급 버그 수정")]`

 * 처리 순서 (우선순위 낮은 순):

     1.  (1, "서버 점검") 또는 (1, "긴급 버그 수정")

     2.  (1, "긴급 버그 수정") 또는 (1, "서버 점검")

     3.  (2, "데이터 백업")

     4.  (3, "보고서 생성")

 * → 반환될 수 있는 결과 (1순위 간 순서는 바뀔 수 있음): `["서버 점검", "긴급 버그 수정", "데이터 백업", "보고서 생성"]` 또는 `["긴급 버그 수정", "서버 점검", "데이터 백업", "보고서 생성"]`



 ### 🤔 어떤 자료구조를 사용하는 것이 좋을까요?

 * 여러 요청 중에서 **항상 우선순위가 가장 높은(숫자가 낮은) 요청을 빠르게 찾아** 제거해야 합니다.

 * 새로운 요청이 추가될 수 있습니다 (이 문제에서는 한 번에 주어지지만, 실제 상황을 가정).

 ➡️ **우선순위 큐(Priority Queue)** 가 이러한 요구사항에 최적화되어 있습니다! 파이썬의 `heapq` 모듈을 사용하면 최소 힙(min-heap)을 기본으로 구현하여, 가장 작은 우선순위 값을 가진 요소를 O(log n) 시간에 효율적으로 찾고 제거할 수 있습니다.



 **비교:**

 * **리스트 사용:** 매번 가장 우선순위가 낮은 요청을 찾으려면 리스트 전체를 검색(O(n))하거나, 리스트를 계속 정렬(O(n log n))해야 해서 비효율적입니다.

 * **우선순위 큐(`heapq`) 사용:** 요청을 힙에 추가하는 데 O(log n), 가장 우선순위 높은 요청을 제거하는 데 O(log n) 시간이 걸려 효율적입니다.

 ### 힌트

 * `import heapq`를 사용하여 heapq 모듈을 가져옵니다.

 * 빈 리스트 `priority_queue = []`를 생성합니다. `heapq`는 이 리스트를 최소 힙으로 사용합니다.

 * 주어진 `requests` 리스트를 순회하면서 각 `(priority, task)` 튜플을 `heapq.heappush(priority_queue, (priority, task))`를 사용하여 힙에 추가합니다. `heapq`는 튜플의 첫 번째 요소(여기서는 `priority`)를 기준으로 힙 속성을 유지합니다.

 * 결과를 저장할 빈 리스트 `processed_order = []`를 만듭니다.

 * `while priority_queue:` 루프를 사용하여 힙이 빌 때까지 반복합니다.

 * 루프 안에서 `priority, task = heapq.heappop(priority_queue)`를 호출하여 가장 작은 우선순위 값을 가진 튜플을 힙에서 제거하고 반환받습니다.

 * 반환받은 `task`를 `processed_order` 리스트에 추가합니다.

 * 루프가 끝나면 `processed_order` 리스트를 반환합니다.

 ### 정답 코드

````python
# %%
import heapq # 이미 위에서 import 했지만, 명확성을 위해 다시 표기

def process_requests_by_priority(requests: list[tuple[int, str]]) -> list[str]:
    """요청들을 우선순위가 낮은 순서대로 처리하여 요청 내용 리스트 반환 (heapq 활용)"""

    priority_queue = [] # heapq 모듈은 이 리스트를 최소 힙으로 사용

    # 모든 요청을 우선순위 큐(최소 힙)에 추가
    # 튜플 (priority, task)를 넣으면 priority 기준으로 힙 구성
    for priority, task in requests:
        heapq.heappush(priority_queue, (priority, task))

    # 처리된 요청 내용을 순서대로 저장할 리스트
    processed_order = []

    # 큐(힙)가 빌 때까지 가장 우선순위가 낮은 요청(힙의 루트)을 꺼냄
    while priority_queue:
        # heappop은 가장 작은 요소(튜플)를 반환하고 힙에서 제거
        priority, task = heapq.heappop(priority_queue)
        processed_order.append(task) # 요청 내용만 저장

    return processed_order

# 테스트 케이스
requests1 = [(3, "보고서 생성"), (1, "서버 점검"), (2, "데이터 백업"), (1, "긴급 버그 수정")]
print(f"요청: {requests1}")
# 1순위인 "서버 점검", "긴급 버그 수정"의 상대적 순서는 실행 시 달라질 수 있음
print(f"-> 처리 순서: {process_requests_by_priority(requests1)}")

requests2 = [(5, "A"), (2, "B"), (4, "C"), (2, "D"), (3, "E")]
print(f"요청: {requests2}")
# 2순위인 B와 D의 상대적 순서는 보장되지 않음
print(f"-> 처리 순서: {process_requests_by_priority(requests2)}")

````

 ## 📚 정렬 (Sorting) 이란? 📈📉



 * **정의**: 여러 개의 데이터 요소들을 특정 **기준(key)** 에 따라

   **일정한 순서(오름차순, 내림차순 등)** 로 재배열하는 작업입니다.

 * **왜 필요할까?**:

     * **탐색 효율 증대**: 정렬된 데이터는 검색이 훨씬 빠릅니다.

       (예: 이진 탐색 O(log n) vs 순차 탐색 O(n)) 🚀

     * **데이터 분석 용이**: 순서대로 나열하면 최솟값, 최댓값,

       데이터 분포 등을 파악하기 쉽고 시각화에도 유리합니다. 📊

     * **알고리즘의 전처리**: 특정 알고리즘은 정렬된 데이터를

       입력으로 요구하는 경우가 많습니다.

     * **사용자 편의성**: 정보를 이름순, 날짜순 등으로 정렬하여 보여주면

       사용자가 이해하고 사용하기 편리합니다.



 * **파이썬 정렬의 특징 (팀소트 Timsort)**:

     * **시간 복잡도**: 평균적으로나 최악의 경우 모두 **O(n log n)** 입니다.

       매우 효율적입니다! ⚡️

     * **공간 복잡도**: 데이터 특성에 따라 다르지만, 추가 메모리가

       필요할 수 있습니다. (최대 O(n))

     * **안정성 (Stability)**: 파이썬의 내장 정렬은 **안정 정렬(Stable Sort)** 입니다.

       즉, **값이 같은 요소들의 상대적인 순서**가 정렬 후에도 **유지**됩니다.

       이 특징은 다중 조건 정렬 시 매우 유용합니다. 👍



 * *(참고)* 과거에는 버블, 선택, 삽입, 병합, 퀵 정렬 등 다양한 알고리즘을

   직접 구현하기도 했지만, 현대 프로그래밍에서는 대부분 언어에 내장된

   고효율 정렬 기능을 사용하는 것이 일반적입니다. 파이썬은 '팀소트'라는

   삽입 정렬과 병합 정렬의 장점을 결합한 알고리즘을 사용합니다.

 ## 🐍 파이썬 내장 정렬 기능 (`sort()`, `sorted()`)



 파이썬은 두 가지 주요 내장 정렬 방법을 제공합니다.

 상황에 맞게 선택하여 사용하는 것이 중요합니다.



 **1. `list.sort()` 메서드**

   * 리스트(list) 객체 **자체**를 **제자리에서(in-place)** 정렬합니다.

   * 즉, **원본 리스트가 직접 변경**됩니다. ♻️

   * 반환값은 `None` 입니다. (정렬된 리스트를 반환하지 않음!)

   * 리스트 타입에서만 사용할 수 있습니다.

````python
# %%
# list.sort() 기본 사용 (오름차순)
my_list = [5, 1, 4, 2, 3]
print(f"정렬 전 리스트: {my_list}")

return_value = my_list.sort() # 제자리 정렬 실행
print(f"my_list.sort() 실행 후 리스트: {my_list}")
print(f"my_list.sort()의 반환값: {return_value}") # None 반환
````

````python
# %%
# list.sort() 내림차순 정렬 (reverse=True)
my_list = [5, 1, 4, 2, 3]
print(f"\n다시 정렬 전 리스트: {my_list}")
my_list.sort(reverse=True) # 내림차순으로 제자리 정렬
print(f"내림차순 정렬 후 리스트: {my_list}")
````

 **2. `sorted()` 내장 함수**

   * 어떤 **반복 가능한(iterable)** 객체든 (리스트, 튜플, 문자열 등)

     입력으로 받을 수 있습니다.

   * 정렬된 요소들을 담은 **새로운 리스트**를 **반환**합니다. ✨

   * **원본 객체는 변경되지 않습니다.** (중요!)

   * 정렬 결과를 새로운 변수에 저장하여 사용합니다.

````python
# %%
# sorted() 기본 사용 (오름차순)
my_tuple = (5, 1, 4, 2, 3)
print(f"원본 튜플: {my_tuple}")

sorted_list = sorted(my_tuple) # 새 리스트를 반환받음
print(f"sorted(my_tuple) 결과 (새 리스트): {sorted_list}")
print(f"sorted() 실행 후 원본 튜플: {my_tuple}") # 원본 변경 없음
````

````python
# %%
# sorted() 내림차순 정렬 (reverse=True)
my_string = "Python"
print(f"\n원본 문자열: {my_string}")
sorted_chars_desc = sorted(my_string, reverse=True) # 문자 기준 내림차순
print(f"문자열 내림차순 정렬 결과: {sorted_chars_desc}")
````

 ## 🔑 `key` 인자 활용: 사용자 정의 정렬 기준



 `sort()` 메서드와 `sorted()` 함수 모두 `key` 라는 매개변수를 가집니다.

 이 `key` 매개변수에 **함수**를 전달하면, 파이썬은 리스트의 각 요소에

 해당 함수를 적용시킨 **결과값을 기준**으로 정렬을 수행합니다.



 * **`lambda` 함수**가 `key`와 함께 매우 유용하게 사용됩니다.

   간단한 기준 함수를 즉석에서 정의할 수 있습니다.



 * **정렬 원리**: 각 요소 `x`에 대해 `key(x)`를 계산하고, 이 `key(x)`의 값들을

   비교하여 요소 `x`들의 순서를 결정합니다.

 ### `key` 사용 예시 1: 문자열 길이 기준 정렬

````python
# %%
words = ["banana", "apple", "kiwi", "strawberry", "orange"]
print(f"원본 단어 리스트: {words}")

# key=len : 각 단어의 '길이'를 기준으로 정렬
sorted_by_length = sorted(words, key=len)
print(f"길이 기준 오름차순 정렬: {sorted_by_length}")

# key=len, reverse=True : 길이 기준 내림차순
sorted_by_length_desc = sorted(words, key=len, reverse=True)
print(f"길이 기준 내림차순 정렬: {sorted_by_length_desc}")
````

 ### `key` 사용 예시 2: 대소문자 구분 없이 정렬

````python
# %%
mixed_case_words = ["Banana", "apple", "Cherry", "date"]
print(f"\n원본 단어 리스트: {mixed_case_words}")

# 기본 정렬은 대문자가 소문자보다 앞에 옴
print(f"기본 정렬 결과: {sorted(mixed_case_words)}")

# key=str.lower : 각 단어를 소문자로 변환한 값을 기준으로 정렬
sorted_case_insensitive = sorted(mixed_case_words, key=str.lower)
print(f"대소문자 무시 정렬: {sorted_case_insensitive}")
````

 ### `key` 사용 예시 3: 리스트 안 튜플의 특정 요소 기준 정렬

````python
# %%
students = [('Alice', 90), ('Bob', 85), ('Charlie', 95)]
print(f"\n학생 데이터 (이름, 점수): {students}")

# key=lambda x: x[1] : 각 튜플 x의 두 번째 요소(점수)를 기준으로 정렬
sorted_by_score = sorted(students, key=lambda x: x[1])
print(f"점수 기준 오름차순 정렬: {sorted_by_score}")

# key=lambda x: x[0], reverse=True : 첫 번째 요소(이름) 기준 내림차순
sorted_by_name_desc = sorted(students, key=lambda x: x[0], reverse=True)
print(f"이름 기준 내림차순 정렬: {sorted_by_name_desc}")
````

 ### `key` 사용 예시 4: 다중 조건 정렬 (튜플 활용)



 여러 기준으로 정렬하고 싶을 때는 `key` 함수가 **튜플**을 반환하도록 만듭니다.

 파이썬은 튜플을 비교할 때 **첫 번째 요소부터 순서대로 비교**합니다.

 첫 번째 요소가 같으면 두 번째 요소, 그것도 같으면 세 번째 요소를 비교합니다.



 예: `key=lambda x: (기준1, 기준2)` -> 기준1로 먼저 정렬, 기준1이 같으면 기준2로 정렬

````python
# %%
data = [
    ('apple', 5, 'red'),
    ('banana', 2, 'yellow'),
    ('cherry', 8, 'red'),
    ('date', 5, 'brown'),
    ('grape', 5, 'purple')
]
print(f"\n과일 데이터 (이름, 수량, 색상): {data}")

# 1순위: 수량(인덱스 1) 오름차순
# 2순위: 이름(인덱스 0) 오름차순
sorted_multi_1 = sorted(data, key=lambda x: (x[1], x[0]))
print(f"\n수량(오름), 이름(오름) 정렬:\n{sorted_multi_1}")

# 1순위: 색상(인덱스 2) 오름차순
# 2순위: 수량(인덱스 1) 내림차순 (수량에 -1 곱하기)
sorted_multi_2 = sorted(data, key=lambda x: (x[2], -x[1]))
print(f"\n색상(오름), 수량(내림) 정렬:\n{sorted_multi_2}")

````

 ## ❓ 객관식, 단답형 및 서술형 문제



 * 각 문제의 난이도를 ⭐️ ~ ⭐️⭐️⭐️⭐️⭐️ 로 표시했습니다.

 ### 🧐 객관식 문제 1 (난이도: ⭐️⭐️⭐️)



 파이썬의 내장 정렬 함수 `sorted()` 와 리스트 메서드 `list.sort()` 의

 주요 차이점으로 **옳은** 것은 무엇인가요?



 1.  `sorted()`는 제자리 정렬(in-place)을 수행하고, `list.sort()`는

     새로운 정렬된 리스트를 반환한다.

 2.  `sorted()`는 리스트만 정렬할 수 있고, `list.sort()`는

     모든 반복 가능한 객체를 정렬할 수 있다.

 3.  `sorted()`는 새로운 정렬된 리스트를 반환하고, `list.sort()`는

     리스트 자체를 제자리에서 정렬한다.

 4.  `sorted()`는 항상 오름차순으로만 정렬하고, `list.sort()`는

     `reverse=True` 옵션을 사용할 수 있다.

 5.  `sorted()`는 튜플만 정렬할 수 있고, `list.sort()`는 리스트만

     정렬할 수 있다.

 ### 정답 및 해설

 **정답: 3번**



 **해설:**

 * `sorted(iterable)` 함수는 어떤 반복 가능한(iterable) 객체든 입력받아,

   그 요소들을 정렬한 **새로운 리스트**를 반환합니다. 원본 객체는 변경되지 않습니다.

 * `list.sort()` 메서드는 리스트 객체에만 사용할 수 있으며,

   해당 리스트 **자체를 제자리에서(in-place)** 정렬합니다.

   반환값은 `None`이며 원본 리스트가 직접 변경됩니다.



 **다른 선택지 설명:**

 1. 반대로 설명되었습니다.

 2. `sorted()`가 더 범용적입니다 (모든 iterable 가능). `list.sort()`는 리스트 전용입니다.

 4. 둘 다 `reverse=True` 옵션을 사용하여 내림차순 정렬이 가능합니다.

 5. `sorted()`는 튜플을 포함한 모든 iterable을 정렬할 수 있습니다.

 ### 🧐 객관식 문제 2 (난이도: ⭐️⭐️⭐️⭐️)



 다음 코드의 실행 결과로 올바른 것은 무엇인가요?



 ```python

 data = [( 'apple' , 5 ), ( 'banana' , 2 ), ( 'cherry' , 8 ), ( 'date' , 5 )]

 result = sorted(data, key = lambda x: (-x[1], x[0]))

 print(result)

 ```



 1. `[('apple', 5), ('banana', 2), ('cherry', 8), ('date', 5)]`

 2. `[('banana', 2), ('apple', 5), ('date', 5), ('cherry', 8)]`

 3. `[('banana', 2), ('date', 5), ('apple', 5), ('cherry', 8)]`

 4. `[('cherry', 8), ('apple', 5), ('date', 5), ('banana', 2)]`

 5. `[('cherry', 8), ('date', 5), ('apple', 5), ('banana', 2)]`

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 `result = sorted(data, key = lambda x: (-x[1], x[0]))` 코드는

 `data` 리스트를 정렬하여 **새로운 리스트** `result`를 생성합니다.

 `key` 함수는 각 요소 `x`(튜플)에 대해 `(-x[1], x[0])`

 즉, (음수 숫자, 문자열) 형태의 튜플을 반환합니다.



 1.  **첫 번째 기준 ( `-x[1]` )**: 튜플의 두 번째 요소(숫자)에

     음수를 취한 값을 기준으로 **오름차순** 정렬합니다.

     이는 원래 숫자를 기준으로 **내림차순** 정렬하는 것과 같습니다.

     * `('cherry', 8)` -> `-8`

     * `('apple', 5)` -> `-5`

     * `('date', 5)` -> `-5`

     * `('banana', 2)` -> `-2`

     음수 기준 오름차순 정렬 순서: `-8` -> `-5` -> `-5` -> `-2`

     즉, 원래 숫자 기준 내림차순: `cherry` -> (`apple`, `date`) -> `banana` 입니다.



 2.  **두 번째 기준 ( `x[0]` )**: 첫 번째 기준 값(`-x[1]`)이 **같을 경우**,

     튜플의 첫 번째 요소(문자열)를 기준으로 **오름차순** 정렬합니다.

     * `('apple', 5)` 와 `('date', 5)` 는 첫 번째 기준 값이 `-5`로 같습니다.

     * 두 번째 기준인 문자열 'apple'과 'date'를 비교하면 'apple'이 'date'보다 앞에 옵니다.



 따라서 최종 정렬 순서는

 `('cherry', 8)`, `('apple', 5)`, `('date', 5)`, `('banana', 2)` 이 됩니다.

 ### 🧐 객관식 문제 3 (난이도: ⭐️⭐️⭐️)



 파이썬의 내장 정렬(팀소트)은 "안정 정렬(Stable Sort)"입니다.

 안정 정렬의 특성으로 올바른 설명은 무엇인가요?



 1. 정렬 속도가 항상 가장 빠르다는 것을 의미한다.

 2. 정렬 시 추가적인 메모리를 전혀 사용하지 않는다는 것을 의미한다.

 3. 어떤 입력 데이터에 대해서도 항상 O(n log n) 시간 복잡도를 보장한다는 것을 의미한다.

 4. 값이 같은 요소들의 상대적인 순서가 정렬 후에도 유지된다는 것을 의미한다.

 5. 역순으로 정렬된 데이터에 대해 가장 효율적이라는 것을 의미한다.

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 **안정 정렬(Stable Sort)** 의 가장 중요한 특징은

 정렬 과정에서 **값이 동일한 요소들의 원래 상대적 순서가

 정렬 후에도 그대로 보존된다**는 점입니다.



 예를 들어, (점수, 이름) 데이터 `[(80, 'Kim'), (90, 'Lee'), (80, 'Park')]` 를

 점수 기준으로 안정 정렬하면 `[(80, 'Kim'), (80, 'Park'), (90, 'Lee')]` 가 됩니다.

 점수가 80점으로 같은 'Kim'과 'Park'의 순서가

 정렬 전('Kim'이 'Park'보다 앞)과 정렬 후에도 동일하게 유지됩니다.

 이는 다중 조건 정렬 시 매우 유용합니다.



 **다른 선택지 설명:**

 1. 안정성 자체는 속도와 직접적인 관련이 없습니다.

 2. 안정 정렬 알고리즘 중에서도 추가 메모리를 사용하는 경우가 있습니다(예: 병합 정렬).

 3. 시간 복잡도 보장은 안정성과는 다른 개념입니다 (팀소트는 둘 다 해당되긴 함).

 5. 특정 데이터 패턴에 대한 효율성은 안정성과는 직접 관련이 없습니다.

 ### ✍️ 단답형 문제 1 (난이도: ⭐️⭐️)



 리스트 `my_data = [10, -5, 3, -8, 0]` 를

 각 숫자의 **절대값**을 기준으로 **내림차순** 정렬한 결과를 작성하세요.



 **힌트:** `key` 인자와 `abs()` 함수, `reverse` 인자를 활용하세요.

 ### 정답 및 해설

 **정답:**

 `[10, -8, -5, 3, 0]`



 **해설:**

 1. 각 숫자의 절대값을 계산합니다:

    `10 -> 10`, `-5 -> 5`, `3 -> 3`, `-8 -> 8`, `0 -> 0`

    절대값 리스트: `[10, 5, 3, 8, 0]`



 2. 이 절대값을 기준으로 내림차순 정렬합니다:

    `10 > 8 > 5 > 3 > 0`



 3. 이 순서에 해당하는 원래 숫자들을 배열합니다:

    `10`, `-8`, `-5`, `3`, `0`



 **파이썬 코드:**

 ```python

 my_data = [10, -5, 3, -8, 0]

 sorted_by_abs_desc = sorted(my_data, key=abs, reverse=True)

 print(sorted_by_abs_desc) # 출력: [10, -8, -5, 3, 0]

 ```

 `key=abs` 는 각 요소의 절대값을 기준으로 사용하겠다는 의미입니다.

 `reverse=True` 는 내림차순 정렬을 의미합니다.

 ## 💻 코드 실습: 시간대별/중요도별 로그 리포트 생성



 상황:

 시스템 장애 분석을 위해, 지난 1시간 동안 수집된 전체 로그 파일 (logs.txt)의 내용을 분석해야 합니다.

 분석가는 로그를 중요도 (Error > Warn > Info 순)와

 발생 시각(오래된 순)에 따라 완전히 정렬된 형태로 보고 싶어 합니다.

 이렇게 정렬된 로그 리포트를 생성하여 문제의 원인을 시간 순서대로 추적하고,

 중요한 이벤트를 먼저 확인하려고 합니다.

 (만약 로그가 실시간으로 들어오고 가장 시급한 로그만 즉시 처리해야 한다면

 우선순위 큐(힙)를 고려할 수 있지만,

 여기서는 이미 수집된 데이터 전체를 정렬하여 리포트를 만드는 것이 목표입니다.)

 ### 상황 설정



 여러 서비스에서 발생한 로그 메시지들이 리스트 `logs` 에 저장되어 있습니다.

 각 로그는 딕셔너리 형태로, 발생 시각(`timestamp`), 로그 레벨(`level`),

 그리고 메시지 내용(`message`)을 포함합니다.



 **로그 데이터:**

 ```python

 logs = [

     {'timestamp': '2025-04-14T11:35:01Z', 'level': 'INFO', 'message': 'User logged in'},

     {'timestamp': '2025-04-14T11:34:58Z', 'level': 'WARN', 'message': 'Disk space low'},

     {'timestamp': '2025-04-14T11:36:05Z', 'level': 'INFO', 'message': 'Data processed successfully'},

     {'timestamp': '2025-04-14T11:34:59Z', 'level': 'ERROR', 'message': 'Database connection failed'},

     {'timestamp': '2025-04-14T11:35:01Z', 'level': 'WARN', 'message': 'API rate limit approaching'},

 ]

 ```



 **최종 목표:** 로그를 다음 **다중 기준**으로 정렬합니다.

 1.  **로그 레벨(level)**: 'ERROR' > 'WARN' > 'INFO' 순서 (심각도 높은 순)

 2.  **타임스탬프(timestamp)**: 같은 레벨 내에서는 시간 순 (오래된 순 -> 최신 순)

 ### Step 1: 타임스탬프 기준으로 정렬하기



 먼저 가장 기본적인 정렬 기준인 **타임스탬프(timestamp)** 를 기준으로

 로그를 **오름차순(시간 순)** 으로 정렬해 봅시다.



 **요구사항:**

 * `logs` 리스트를 `sorted()` 함수를 사용하여 정렬하세요.

 * `key` 인자에 람다 함수를 사용하여 각 로그 딕셔너리의 `'timestamp'` 값을

   정렬 기준으로 삼으세요.

 * 타임스탬프 문자열은 그대로 비교해도 시간 순 정렬이 가능합니다 (ISO 형식).



 **실행 코드:** (아래 셀에 코드를 작성하고 실행해보세요)

 ### 💡 Step 1 힌트

 * `sorted()` 함수의 첫 번째 인자로는 정렬할 대상인 `logs` 리스트를 전달합니다.

 * `key` 인자에는 각 요소(로그 딕셔너리)를 입력받아 정렬의 기준이 될 값을

   반환하는 함수를 지정합니다.

 * `lambda log: log['timestamp']` 는 딕셔너리 `log` 를 받아서

   `'timestamp'` 키의 값(시간 문자열)을 반환하는 익명 함수입니다.

 * ISO 8601 형식의 타임스탬프 문자열은 문자열 비교만으로도

   시간 순서대로 정렬됩니다.

 ### ✅ Step 1 Solution / Code

````python
# %%
# Step 1 Solution Code
logs_data_step1 = [
    {'timestamp': '2025-04-14T11:35:01Z', 'level': 'INFO', 'message': 'User logged in'},
    {'timestamp': '2025-04-14T11:34:58Z', 'level': 'WARN', 'message': 'Disk space low'},
    {'timestamp': '2025-04-14T11:36:05Z', 'level': 'INFO', 'message': 'Data processed successfully'},
    {'timestamp': '2025-04-14T11:34:59Z', 'level': 'ERROR', 'message': 'Database connection failed'},
    {'timestamp': '2025-04-14T11:35:01Z', 'level': 'WARN', 'message': 'API rate limit approaching'},
]

# key=lambda 를 이용하여 timestamp 값을 기준으로 정렬
solution_step1 = sorted(logs_data_step1, key=lambda log: log['timestamp'])

# 결과 확인용 출력 (위 셀에서 이미 확인했으므로 주석 처리 가능)
# print("--- Step 1 Solution Output ---")
# for log_entry in solution_step1:
#     print(log_entry)
````

 ### Step 2: 로그 레벨과 타임스탬프 다중 기준으로 정렬하기

 * `key` 람다 함수가 **튜플** `(레벨 우선순위, 타임스탬프)` 를 반환하도록

     level_priority = {'ERROR': 0, 'WARN': 1, 'INFO': 2}

     `(level_priority[log['level']], log['timestamp'])` 와 같은

     * `log['timestamp']` : 타임스탬프 문자열 (두 번째 비교 기준)

 3.  **튜플 비교**: 파이썬은 튜플 `(a, b)` 와 `(c, d)` 를 비교할 때,

     먼저 `a`와 `c`를 비교합니다. 만약 `a`와 `c`가 같으면, 그 다음으로 `b`와 `d`를

     비교하여 순서를 결정합니다.

````python
# %%
# Step 2 Solution Code
logs_data_step2 = [
    {'timestamp': '2025-04-14T11:35:01Z', 'level': 'INFO', 'message': 'User logged in'},
    {'timestamp': '2025-04-14T11:34:58Z', 'level': 'WARN', 'message': 'Disk space low'},
    {'timestamp': '2025-04-14T11:36:05Z', 'level': 'INFO', 'message': 'Data processed successfully'},
    {'timestamp': '2025-04-14T11:34:59Z', 'level': 'ERROR', 'message': 'Database connection failed'},
    {'timestamp': '2025-04-14T11:35:01Z', 'level': 'WARN', 'message': 'API rate limit approaching'},
]

# 로그 레벨 우선순위 맵
level_map = {'ERROR': 0, 'WARN': 1, 'INFO': 2}

# key 람다 함수가 (레벨 우선순위, 타임스탬프) 튜플을 반환
solution_step2 = sorted(logs_data_step2, key=lambda log: (level_map[log['level']], log['timestamp']))

# 결과 확인용 출력 (위 셀에서 이미 확인했으므로 주석 처리 가능)
# print("\n--- Step 2 Solution Output ---")
# for log_entry in solution_step2:
#     print(log_entry)
````

 ## 📚 그래프 (Graph) 란? 🌐🔗



 * **정의**: 여러 개의 **점(정점 Vertex 또는 노드 Node)** 들이 서로

   **선(간선 Edge 또는 링크 Link)** 으로 연결된 관계를 표현하는 자료구조입니다.

   (데이터들이 순서대로 나열된 리스트와 달리, 관계 중심으로 얽혀있는 비선형 구조!)



 * **구성 요소**:

     * **정점 (Vertex / Node)**: 데이터를 나타내는 '점'. (예: 사람, 도시, 웹페이지)

     * **간선 (Edge / Link)**: 정점 사이의 '관계'를 나타내는 '선'. (예: 친구 관계, 도로, 웹 링크)



1.png



 * **왜 사용할까?**: 세상의 많은 것들이 **서로 연결된 관계**를 가지고 있어요.

   이런 복잡한 연결 구조를 표현하고 분석하는 데 그래프가 아주 유용합니다!

     * **예시**: 친구 관계망(카카오톡 친구 목록!), 지하철 노선도, 인터넷 웹페이지 연결, 도로 지도 등


 * **그래프 관련 용어 살짝 알기**:

     * **무방향 그래프**: 선(간선)에 방향이 없어요. A와 B가 친구면, B도 A의 친구인 것처럼요.

     * **방향 그래프**: 선(간선)에 화살표처럼 방향이 있어요. A가 B 웹페이지로 링크를 걸었다고 B가 A로 링크를 건 것은 아닌 것처럼요.

     * **가중치 그래프**: 선(간선)마다 숫자 값(가중치/비용)이 있어요. 도시 사이의 거리나 도로 통행료처럼요.

     * **비가중치 그래프**: 선(간선)에 별도 값이 없어요 (모든 연결이 동일하게 중요).

  [링크: 그래프 시각화](https://www.w3schools.com/dsa/dsa_theory_graphs.php)

 ## 📝 그래프를 코드로 표현하는 방법: 인접 리스트!



 그래프의 연결 정보를 컴퓨터가 알아들을 수 있게 저장해야겠죠? 여러 방법이 있지만, 우리는 가장 많이 쓰이는 **인접 리스트(Adjacency List)** 방식을 배울 거예요.



 ### 인접 리스트 (Adjacency List) - 딕셔너리 활용

2.png


 * **아이디어**: "각각의 점(정점)마다, 그 점이랑 직접 연결된 친구(이웃)들이 누구인지 목록으로 알려주자!"

 * **파이썬 구현**: **딕셔너리**를 사용하는 것이 아주 편리해요!

     * **Key**: 각 정점 (예: 'A', 'B', 0, 1 등)

     * **Value**: 해당 정점과 직접 연결된 **이웃 정점들의 리스트 (또는 세트)**

 * **예시**:

   ```

     A -- B

     |    |

     C -- D

   ```

   위 그래프를 인접 리스트로 표현하면:

   ```python

   graph = {

       'A': ['B', 'C'],

       'B': ['A', 'D'],

       'C': ['A', 'D'],

       'D': ['B', 'C']

   }

   ```

   'A'의 이웃은 'B'와 'C'라는 것을 바로 알 수 있죠!



 * **장점**:

     * **메모리 효율**: 딱 필요한 연결 정보만 저장해서, 특히 연결선(간선)이 많지 않은 **희소 그래프(Sparse Graph)** 에 좋아요. 💾🔵

     * **이웃 찾기 빠름**: 특정 정점의 이웃들을 바로 리스트(또는 세트)에서 꺼내볼 수 있어요. ⚡️🔵



 * **(참고) 인접 행렬**: 표(2차원 배열)를 이용하는 방법도 있지만, 연결이 적은 그래프에선 메모리 낭비가 심해서 인접 리스트를 더 많이 써요.

 3.png



 * **파이썬 구현 예시 (딕셔너리 & `defaultdict`)**:

````python
# %%
# 파이썬으로 인접 리스트 만들기

# 예제 그래프:
#   1 --- 2
#   | \   |
#   |  \  |
#   0 --- 3 --- 4

from collections import defaultdict

# defaultdict(list) : 키가 없을 때 자동으로 빈 리스트를 값으로 만들어줌 (편리!)
adj_list = defaultdict(list)
print(f"빈 인접 리스트 (defaultdict): {adj_list}")

# 간선 정보 (정점 쌍 리스트)
edges = [(1, 2), (1, 0), (1, 3), (0, 3), (2, 3), (3, 4)]
print(f"\n주어진 간선들: {edges}")

# 인접 리스트 채우기 (무방향 그래프이므로 양쪽 모두 추가)
for u, v in edges:
    adj_list[u].append(v) # u의 이웃으로 v 추가
    adj_list[v].append(u) # v의 이웃으로 u 추가

print("\n인접 리스트 완성:")
# 보기 좋게 일반 딕셔너리로 변환 후 정렬해서 출력
for node in sorted(adj_list.keys()):
    print(f"  정점 {node}의 이웃: {sorted(adj_list[node])}")

# 특정 노드의 이웃 확인하기
node_to_check = 3
print(f"\n정점 {node_to_check}의 이웃들: {sorted(adj_list[node_to_check])}")
````

 ## 🤔 BFS vs DFS 간단 맛보기 ("넓게" vs "깊게")



 그래프를 만들었으니, 이제 그래프 위를 돌아다니는(탐색하는) 방법을 알아야겠죠?

 대표적인 두 가지 방법 **BFS**와 **DFS**의 컨셉만 살짝 맛봅시다!

 4.png



 * **BFS (너비 우선 탐색)**:

     * "가까운 곳부터 차근차근, 옆으로 넓게!"

     * 시작점에서 가까운 순서(레벨 1 -> 레벨 2 -> ...)대로 탐색해요.

     * 마치 물결이 퍼져나가는 모습과 같아요.

     * **최단 거리** 찾을 때 유리해요.



 * **DFS (깊이 우선 탐색)**:

     * "한 길로 끝까지 파고들어 보자!"

     * 한 방향으로 갈 수 있는 데까지 깊숙이 들어갔다가, 막히면 돌아 나와 다른 길로 가요.

     * 마치 미로 찾기 할 때 한쪽 벽만 따라가는 느낌이에요.

     * **경로의 존재 여부** 확인 등에 쓰여요.

 ## 📚 너비 우선 탐색 (Breadth-First Search, BFS) 란? 🌊➡️

 * **정의**: 그래프 탐색 알고리즘 중 하나로, 시작 정점(Vertex)에서 **가까운 정점부터** 순서대로, 즉 **넓게(너비 우선)** 탐색하는 방법.

 * **핵심 아이디어**: 시작 노드에서 출발하여, **거리가 1인 이웃**들을 모두 방문하고, 그 다음 **거리가 2인 이웃**들을 모두 방문하는 방식으로, 마치 물결이 퍼져나가듯 탐색을 진행. 🌊

 * **주요 특징**:

     * **큐(Queue)** 자료구조를 사용하여 구현 (FIFO: First-In, First-Out).

     * **최단 경로 보장 (가중치 없는 그래프)**: 시작 노드에서 특정 노드까지의 가장 적은 간선 수를 거치는 경로를 찾을 수 있음. 🥇 (깊이 우선 탐색(DFS)은 보장 안 함)

     * 재귀적으로 동작하지 않음 (일반적으로 반복문과 큐 사용).

 * **비유**:

     * 연못에 돌 던지기: 돌이 떨어진 지점(시작 노드)에서 가장 가까운 물결(레벨 1 이웃)이 먼저 퍼지고, 그 다음 물결(레벨 2 이웃)이 퍼져나가는 모습.

     * 친구 찾기: 내 바로 옆 친구들(레벨 1)에게 먼저 물어보고, 그 친구들의 친구들(레벨 2)에게 물어보는 방식.

 * *[시각 자료 삽입: BFS 탐색 과정을 단계별로 보여주는 애니메이션 또는 그림]*

 ## 🤔 BFS (너비 우선 탐색) 핵심 아이디어: 넓게 펼치며 탐색!



 BFS는 마치 연못에 돌을 던졌을 때 **물결이 퍼져나가듯** 동작하는 탐색 방법입니다.



 1.  **시작점**에서 가장 **가까운 곳**부터 방문합니다 (거리 1).

 2.  그다음, **조금 더 먼 곳**들을 방문합니다 (거리 2).

 3.  이런 식으로 **점점 더 넓은 영역**으로, **층(Level)별로** 탐색을 진행합니다.



 이 방식 덕분에 BFS는 **최단 경로** (간선의 수가 가장 적은 경로)를 찾는 데

 매우 유용하게 사용됩니다. (가중치가 없는 그래프에서)

 ---



 ## ⚙️ BFS 동작 원리: 큐(Queue)가 핵심!



 BFS가 층(Level)별 탐색을 할 수 있는 비결은 바로 **큐(Queue)** 자료구조를

 사용하기 때문입니다. 큐는 **먼저 들어온 것이 먼저 나가는(FIFO)** 특징이 있죠.



 1.  **시작**:

     * 탐색 **시작 노드**를 **큐**에 넣습니다.

     * "이미 방문했다"는 표시를 남깁니다 (`visited`).



 2.  **탐색 (큐가 빌 때까지 반복)**:

     * 큐에서 **가장 먼저 들어왔던 노드**를 꺼냅니다 (`current`).

     * `current` 노드를 **처리**합니다 (예: 화면에 출력).

     * `current` 노드에 **연결된 이웃 노드**들을 살펴봅니다.

     * 이웃 노드 중 **아직 방문 안 한 노드**가 있다면:

         * "방문했다"고 **표시**하고,

         * **큐에 넣습니다**. (다음 레벨 탐색 대상)



 3.  **종료**: 큐가 비면, 시작 노드에서 갈 수 있는 모든 노드를 방문한 것입니다.



 **핵심**: 큐에 먼저 들어간 노드(가까운 노드)가 먼저 나오고, 그 이웃들이 큐의

 뒤쪽에 추가됩니다. 따라서 자연스럽게 가까운 레벨부터 탐색하게 됩니다. 👍

 ---



 ## 🐍💻 BFS 간단 구현 (Python - `collections.deque` 활용)



 그래프는 **인접 리스트 (딕셔너리)** 로 표현합니다.

 `{노드: [이웃1, 이웃2, ...]}`

````python
# %%
from collections import deque

def bfs_simple(graph, start_node):
    """너비 우선 탐색(BFS)의 방문 순서를 반환하는 간단한 함수"""
    # 시작 노드가 그래프에 있는지 확인
    if start_node not in graph:
        return []

    visited = set()           # 방문 기록용 집합
    queue = deque([start_node]) # 탐색 대기열 큐
    visited.add(start_node)   # 시작 노드 방문 처리

    visited_order = [] # 방문 순서 기록용 리스트

    while queue: # 큐에 탐색할 노드가 남아있는 동안
        # 큐의 맨 앞에서 노드를 꺼냄
        current_node = queue.popleft()
        visited_order.append(current_node) # 방문 순서 기록

        # 현재 노드의 이웃들을 확인
        for neighbor in graph.get(current_node, []): # .get()으로 키 없는 경우 처리
            # 아직 방문하지 않은 이웃이라면
            if neighbor not in visited:
                visited.add(neighbor)   # 방문 표시
                queue.append(neighbor) # 큐에 추가 (다음 탐색 대상)

    return visited_order

# 예제 그래프
graph_example = {
    'A': ['B', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['B'],
    'D': ['A', 'E'],
    'E': ['B', 'D']
}

# BFS 실행
bfs_result = bfs_simple(graph_example, 'A')
print(f"그래프: {graph_example}")
print(f"시작 노드 'A'에서 BFS 방문 순서: {bfs_result}")
# 출력 예시: ['A', 'B', 'D', 'C', 'E'] (같은 레벨 B, D 순서는 달라질 수 있음)
````

 ## ⏱️💾 BFS 복잡도: 얼마나 빠르고, 메모리는 얼마나 쓸까?



 (V: 노드(Vertex) 개수, E: 간선(Edge) 개수)



 * **시간 복잡도: O(V + E)** 🚀

   * 모든 노드(V)를 딱 한 번씩 방문하고,

   * 모든 간선(E)을 딱 한 번씩 검사하기 때문입니다.

   * 그래프 크기에 비례하는 아주 효율적인 시간!



 * **공간 복잡도: O(V)** 💾

   * **큐(Queue)** 에 최악의 경우 모든 노드(V)가 들어갈 수 있습니다.

   * **방문 기록(visited)** 도 모든 노드(V) 정보를 저장해야 합니다.

   * 필요한 메모리 공간도 노드 수에 비례합니다.



 **결론**: BFS는 그래프를 탐색하는 매우 **빠르고 효율적인** 방법입니다!

 ## 📍🗺️ BFS 활용 사례



 BFS는 특정 조건 하에서 매우 유용하게 사용됩니다.



 1.  **최단 경로 찾기 (가중치 없는 그래프)**:

     * BFS는 시작 노드로부터 **가장 가까운 노드부터** 탐색하므로, 특정 목표 노드에 도달했을 때 그 경로는 항상 **간선의 수가 가장 적은 최단 경로**임이 보장됩니다.

     * 예: 소셜 네트워크에서 두 사람 사이의 최소 연결 단계 찾기, 미로 찾기 게임에서 최단 탈출 경로 찾기.

     * (구현 Tip: 각 노드까지의 거리를 기록하는 배열/딕셔너리를 함께 관리)



 2.  **연결 요소 (Connected Components) 찾기**:

     * 무방향 그래프에서 서로 연결된 노드들의 집합(연결 요소)을 찾는 데 사용될 수 있습니다.

     * 방문하지 않은 노드에서 BFS를 시작하여 도달 가능한 모든 노드를 찾으면 하나의 연결 요소가 됩니다. 이를 모든 노드를 방문할 때까지 반복합니다.



 3.  **네트워크 브로드캐스팅 시뮬레이션**:

     * 네트워크 상에서 메시지가 시작 노드로부터 주변 노드로 퍼져나가는 과정을 시뮬레이션할 때 BFS의 레벨별 탐색 방식이 유용합니다.



 4.  **웹 크롤러 (기본)**:

     * 특정 웹 페이지에서 시작하여 링크를 따라 연결된 페이지들을 레벨별로 방문하며 수집하는 기본적인 크롤러 구현에 활용될 수 있습니다.



 5.  **그래프 관련 문제 해결**:

     * 사이클 존재 여부 확인 (약간의 수정 필요)

     * 이분 그래프(Bipartite Graph) 판별 등 다양한 그래프 알고리즘의 기초로 사용됩니다.

 ## ✨ AI 분야에서의 BFS 응용 (소개)



 BFS의 기본 원리는 AI의 다양한 분야에서도 응용될 수 있습니다.



 * **이미지 세그멘테이션 (Image Segmentation)**:

     * 이미지 내에서 특정 픽셀(시작점)과 연결된 동일한 속성(색상, 질감 등)을 가진 **인접 픽셀 영역**을 찾는 데 BFS를 응용할 수 있습니다. (Flood fill 알고리즘과 유사)

 * **추천 시스템 (Recommendation Systems)**:

     * 사용자-아이템 관계 그래프에서 특정 사용자와 연결된 친구들(레벨 1), 그 친구들의 친구들(레벨 2)을 탐색하며 **관심사나 선호도가 유사한 사용자/아이템**을 찾는 데 활용될 수 있습니다.

 * **상태 공간 탐색 (State-Space Search)**:

     * 게임이나 퍼즐 문제에서 현재 상태(노드)에서 가능한 다음 상태(이웃 노드)로 이동하며 목표 상태까지의 **최단 해결 경로**를 찾는 데 BFS가 사용될 수 있습니다. (예: 루빅스 큐브 해법 찾기 - 상태 공간 매우 큼)

 * **로봇 경로 계획 (Path Planning)**:

     * 로봇이 이동할 수 있는 공간을 격자(grid) 그래프로 표현하고, 시작 지점에서 목표 지점까지 장애물을 피해 가는 **최단 경로**를 찾는 데 BFS를 활용할 수 있습니다.



 * **핵심**: BFS는 **가장 가까운 관계** 또는 **최단 거리(단계)** 가 중요한 문제 상황에서 유용하게 적용될 수 있는 기본적인 탐색 도구입니다.

 ### 🧐 객관식 문제 1 (난이도: ⭐️⭐️)



 너비 우선 탐색(BFS) 알고리즘에 대한 설명으로 **가장 적절한** 것은 무엇인가요?



 1.  시작 노드에서 가장 멀리 있는 노드부터 탐색한다.

 2.  스택(Stack) 자료구조를 주로 사용하여 구현한다.

 3.  가중치가 있는 그래프에서 항상 최단 경로를 찾는다.

 4.  시작 노드에서 가까운 노드부터 레벨(level) 단위로 탐색한다.

 5.  시간 복잡도는 항상 O(V²)이다.

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 BFS(너비 우선 탐색)는 시작 노드를 기준으로 거리가 1인 이웃, 거리가 2인 이웃 등 **가까운 노드부터 순서대로, 같은 레벨(거리)에 있는 노드들을 우선적으로 탐색**하는 방식입니다. 이는 마치 물결이 퍼져나가는 모습과 같습니다.

 1. BFS는 가까운 노드부터 탐색합니다.

 2. BFS는 큐(Queue)를 사용하여 구현합니다. (스택은 DFS에서 사용)

 3. BFS는 가중치가 없는 그래프에서의 최단 경로(최소 간선 수)를 보장합니다. 가중치가 있는 그래프의 최단 경로는 다익스트라 알고리즘 등을 사용합니다.

 5. 인접 리스트 사용 시 시간 복잡도는 O(V + E)입니다.

 ### 🧐 객관식 문제 2 (난이도: ⭐️⭐️⭐️)



 BFS 알고리즘 구현 시 **큐(Queue)** 자료구조를 사용하는 주된 이유는 무엇인가요?



 1.  가장 최근에 방문한 노드의 이웃을 먼저 탐색하기 위해 (LIFO)

 2.  방문한 노드를 효율적으로 저장하고 중복 방문을 막기 위해

 3.  노드들을 우선순위에 따라 정렬하여 처리하기 위해

 4.  먼저 발견된(시작 노드에서 가까운) 노드들을 먼저 처리하여 레벨별 탐색 순서를 유지하기 위해 (FIFO)

 5.  그래프의 간선 정보를 효율적으로 저장하기 위해

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 큐는 **FIFO(First-In, First-Out)** 특성을 가집니다. BFS에서는 시작 노드에서 가까운(먼저 발견된) 노드들을 큐에 먼저 넣습니다. 큐의 FIFO 특성 덕분에, 큐에서 노드를 꺼내 처리할 때 항상 현재 탐색 중인 레벨의 노드들이 먼저 나오게 되고, 그 다음 레벨의 노드들은 나중에 처리됩니다. 이는 BFS의 핵심인 **레벨별 탐색 순서**를 자연스럽게 구현하도록 돕습니다.

 1. LIFO는 스택의 특징이며 DFS 방식입니다.

 2. 방문 기록은 보통 set이나 배열을 사용합니다.

 3. 우선순위 처리는 우선순위 큐를 사용합니다.

 5. 간선 정보 저장은 인접 행렬이나 인접 리스트를 사용합니다.

 ### 🧐 객관식 문제 3 (난이도: ⭐️⭐️⭐️⭐️)



 다음 그래프에서 노드 'A'를 시작으로 BFS를 수행할 때, 방문 순서로 **가능한** 것은 무엇인가요? (단, 같은 레벨의 이웃 노드 방문 순서는 알파벳 순서라고 가정)



 ```

   A --- B

   | \   |

   C---D-E



 1. A, B, C, D, E

 2. A, C, B, D, E

 3. A, B, E, C, D

 4. A, D, C, B, E

 5. A, B, D, E, C

 ### 정답 및 해설

 **정답: 1번**



 **해설:**

 BFS는 레벨(거리) 순서대로 탐색하며, 큐(Queue)를 사용하여 탐색할 노드를 관리합니다.

 이웃 노드를 큐에 추가할 때는 문제의 가정에 따라 **알파벳 오름차순**으로 넣습니다.



 1.  **시작**:

     * 큐에 시작 노드 'A'를 넣고 방문 처리합니다.

     * `queue = deque(['A'])`

     * `visited = {'A'}`

     * `방문 순서 = []`



 2.  **Level 0 처리**:

     * 'A'를 큐에서 꺼냅니다 (`popleft`). 방문 순서에 추가: `[A]`.

     * 'A'의 이웃 ['B', 'C', 'D']를 확인합니다.

     * 모두 방문하지 않았으므로, **알파벳 순서대로** 큐에 넣고 방문 처리합니다.

     * `queue = deque(['B', 'C', 'D'])`

     * `visited = {'A', 'B', 'C', 'D'}`



 3.  **Level 1 처리 (첫 번째 노드 'B')**:

     * 'B'를 큐에서 꺼냅니다. 방문 순서에 추가: `[A, B]`.

     * 'B'의 이웃 ['A', 'E']를 확인합니다.

     * 'A'는 이미 방문했습니다.

     * 'E'는 방문하지 않았으므로, 큐에 넣고 방문 처리합니다.

     * `queue = deque(['C', 'D', 'E'])`

     * `visited = {'A', 'B', 'C', 'D', 'E'}`



 4.  **Level 1 처리 (두 번째 노드 'C')**:

     * 'C'를 큐에서 꺼냅니다. 방문 순서에 추가: `[A, B, C]`.

     * 'C'의 이웃 ['A', 'D']를 확인합니다.

     * 'A', 'D' 모두 이미 방문했습니다. (큐에 추가할 노드 없음)

     * `queue = deque(['D', 'E'])`



 5.  **Level 1 처리 (세 번째 노드 'D')**:

     * 'D'를 큐에서 꺼냅니다. 방문 순서에 추가: `[A, B, C, D]`.

     * 'D'의 이웃 ['A', 'C', 'E']를 확인합니다.

     * 'A', 'C', 'E' 모두 이미 방문했습니다. (큐에 추가할 노드 없음)

     * `queue = deque(['E'])`



 6.  **Level 2 처리 (첫 번째 노드 'E')**:

     * 'E'를 큐에서 꺼냅니다. 방문 순서에 추가: `[A, B, C, D, E]`.

     * 'E'의 이웃 ['B', 'D']를 확인합니다.

     * 'B', 'D' 모두 이미 방문했습니다. (큐에 추가할 노드 없음)

     * `queue = deque([])`



 7.  **종료**: 큐가 비었으므로 탐색을 종료합니다.



 최종 방문 순서는 `A, B, C, D, E` 입니다. 따라서 **1번**이 올바른 답입니다.



 **다른 선택지가 틀린 이유:**

 * 2번, 4번: 시작 노드 'A'의 이웃을 알파벳 순서('B', 'C', 'D')대로 큐에 넣지 않았습니다.

 * 3번, 5번: 레벨 2에 해당하는 'E' 노드가 레벨 1의 'C' 또는 'D'보다 먼저 방문되었습니다. BFS는 레벨 순서대로 탐색해야 합니다.

 ### ✍️ 단답형 / 서술형 문제 1 (난이도: ⭐️⭐️⭐️)



 **가중치가 없는(Unweighted)** 그래프에서 두 노드 간의 **최단 경로(Shortest Path)** 를 찾을 때 BFS 알고리즘이 DFS(깊이 우선 탐색)보다 선호되는 이유는 무엇인가요?

 ### 정답 및 해설

 **정답:**



 BFS(너비 우선 탐색)는 시작 노드로부터 **가까운 거리(레벨) 순서대로** 노드를 탐색하기 때문입니다. BFS는 시작 노드에서 거리가 1인 모든 노드를 방문하고, 그 다음 거리가 2인 모든 노드를 방문하는 방식으로 진행됩니다. 따라서 BFS 과정에서 특정 목표 노드를 **처음 발견했을 때**, 그 경로는 시작 노드로부터 해당 목표 노드까지 **가장 적은 수의 간선**을 거쳐 도달한 경로가 됩니다. 즉, 가중치가 없는 그래프에서는 간선의 수가 경로의 길이(비용)가 되므로, BFS는 자연스럽게 최단 경로를 찾게 됩니다.



 반면, DFS(깊이 우선 탐색)는 한 경로를 따라 최대한 깊이 들어갔다가 막히면 되돌아 나오는 방식입니다. 따라서 DFS가 목표 노드를 처음 발견하더라도 그 경로가 최단 경로라는 보장이 없습니다. 더 깊이 들어갔다가 돌아 나와서 찾은 경로일 수도 있기 때문입니다. DFS로 최단 경로를 찾으려면 모든 가능한 경로를 탐색하고 길이를 비교해야 할 수 있어 비효율적입니다.

 ### ✍️ 단답형 / 서술형 문제 2 (난이도: ⭐️⭐️⭐️⭐️)



 BFS 알고리즘을 구현할 때, **방문한 노드를 기록하는 것(`visited` set 또는 배열 사용)** 이 왜 필수적인지 설명하고, 만약 방문 기록을 하지 않으면 어떤 문제가 발생할 수 있는지 설명하세요.

 ### 정답 및 해설

 **정답:**



 **방문 기록이 필수적인 이유:**

 BFS 알고리즘에서 방문한 노드를 기록하는 것은 **불필요한 중복 탐색을 방지**하고 **무한 루프(Infinite Loop)** 에 빠지는 것을 막기 위해 필수적입니다.



 **방문 기록을 하지 않을 경우 발생할 수 있는 문제:**



 1.  **무한 루프**: 만약 그래프에 **사이클(Cycle)** 이 존재하는 경우 (예: A -> B -> C -> A), 방문 기록이 없다면 BFS는 사이클 내의 노드들을 계속해서 큐에 넣고 방문하는 과정을 무한히 반복하게 됩니다. 예를 들어 A를 방문하고 B를 큐에 넣고, B를 방문하고 C를 큐에 넣고, C를 방문하고 A를 큐에 넣으면, 다시 A를 방문하고 B를 큐에 넣는 과정이 끝나지 않습니다.

 2.  **비효율적인 탐색**: 사이클이 없더라도, 특정 노드가 여러 다른 노드로부터 이웃으로 연결되어 있을 수 있습니다. 방문 기록이 없다면, 이미 큐에 추가되었거나 처리된 노드가 다른 경로를 통해 다시 큐에 추가될 수 있습니다. 이는 동일한 노드와 그 이웃들을 여러 번 중복해서 탐색하게 만들어 알고리즘의 시간 복잡도를 O(V+E)보다 훨씬 나쁘게 만들고 매우 비효율적인 탐색을 유발합니다.



 따라서 `visited` 자료구조를 사용하여 각 노드를 최대 한 번만 방문하고 큐에 넣도록 보장하는 것은 올바르고 효율적인 BFS 수행을 위해 반드시 필요합니다.

 # BFS를 이용한 최단 거리 찾기 (단계별 학습)



 이 노트북은 너비 우선 탐색(BFS) 알고리즘을 사용하여 가중치가 없는 그래프에서 두 노드 간의 최단 거리를 찾는 방법을 단계별로 학습합니다.

 ### Step 1: 필요한 것들 준비하기



 BFS를 시작하기 전에 몇 가지 준비가 필요합니다.



 1.  `deque`를 import 합니다. 파이썬의 `collections` 모듈에서 `deque`를 가져와 효율적인 큐(Queue) 자료구조로 사용합니다.

 2.  시작 노드와 목표 노드가 같은 경우(거리 0), 시작 노드가 그래프에 없는 경우 등 간단한 예외 상황을 먼저 처리하여 불필요한 탐색을 방지합니다.

 3.  **큐(Queue)** 를 만듭니다. BFS는 큐를 사용하여 탐색할 노드를 관리합니다. 시작 노드와 시작 노드까지의 거리(0) 정보를 튜플 `(start_node, 0)` 형태로 큐에 넣습니다.

 4.  **방문 기록 + 거리 저장**: 어떤 노드를 방문했는지, 그리고 시작 노드로부터 그 노드까지의 최단 거리가 얼마인지 기록할 방법이 필요합니다. 딕셔너리 `distances = {start_node: 0}` 를 사용하면 편리합니다.

     * 딕셔너리에 키(노드)가 **존재하는지 확인**하는 것으로 **방문 여부**를 알 수 있습니다.

     * 딕셔너리의 값으로 해당 노드까지의 **최단 거리**를 저장합니다. BFS의 특성상 특정 노드를 처음 발견했을 때의 거리가 항상 최단 거리입니다.

 #### 예시코드 1: 초기 설정

 아래 코드는 `shortest_path_bfs` 함수의 시작 부분과 Step 1에 해당하는 초기화 과정을 보여줍니다.

````python
# %%
# 예시 코드 1: 초기 설정 부분
from collections import deque

def shortest_path_bfs_step1_example(graph, start_node, end_node):
    # --- Step 1: 필요한 것들 준비하기 ---
    print("--- Step 1: 초기 설정 ---")
    if start_node == end_node: # 시작과 목표가 같으면 거리는 0
        print(f"시작 노드와 목표 노드가 {start_node}로 동일합니다.")
        return 0
    if start_node not in graph: # 시작 노드가 그래프에 없으면 도달 불가
        print(f"오류: 시작 노드 {start_node}가 그래프에 없습니다.")
        return -1

    # 1. 큐 생성 및 시작점 추가: (노드, 거리)
    queue = deque([(start_node, 0)])
    print(f"큐 생성 및 시작 노드 추가: {list(queue)}")

    # 2. 방문 기록 및 거리 저장용 딕셔너리: {노드: 거리}
    distances = {start_node: 0}
    print(f"distances 딕셔너리 초기화: {distances}")

    # (이후 탐색 로직은 생략)
    print("--- 초기 설정 완료 ---")
    # 실제 함수에서는 여기서 탐색을 계속 진행합니다.
    # 이 예시에서는 초기화 부분만 보여주기 위해 None을 반환합니다.
    return None

# 예제 그래프 (테스트용)
graph_ex = {
    # 0: , 1: , 2: , 3:
}
shortest_path_bfs_step1_example(graph_ex, 0, 3)
shortest_path_bfs_step1_example(graph_ex, 0, 0)
shortest_path_bfs_step1_example(graph_ex, 4, 3) # 시작 노드가 없는 경우
````

 ### Step 2: BFS 탐색 시작 (큐가 빌 때까지)



 이제 큐를 이용하여 본격적인 BFS 탐색을 시작합니다. 큐는 다음에 방문해야 할 노드들을 레벨 순서(시작 노드로부터의 거리가 가까운 순서)대로 저장하고 있습니다.



 1.  `while queue:` : 큐에 탐색할 노드가 남아있는 동안 계속 반복합니다. 큐가 비게 되면, 시작 노드에서 연결된 모든 노드를 방문했다는 의미입니다.

 2.  **노드 꺼내기**: `queue.popleft()` 를 사용하여 큐의 가장 앞에 있는 원소, 즉 `(현재 노드, 현재 거리)` 튜플을 꺼냅니다. `popleft()`는 큐의 선입선출(FIFO) 동작을 구현합니다.

 3.  **목표 도착 확인**: 큐에서 꺼낸 `현재 노드`가 우리가 찾던 `end_node` 인지 확인합니다.

     * 만약 `현재 노드`가 `end_node`와 같다면, 탐색에 성공한 것입니다! BFS는 항상 최단 거리를 먼저 찾으므로, 이때의 `현재 거리`가 바로 시작 노드로부터 목표 노드까지의 최단 거리입니다. 이 값을 반환하고 함수를 종료합니다.

 #### 예시코드 2: 큐 처리 및 목표 확인

 아래 코드는 `while` 루프의 시작 부분과 큐에서 노드를 꺼내 목표 노드인지 확인하는 과정을 보여줍니다.

````python
# %%
# 예시 코드 2: 큐 처리 및 목표 확인 부분
from collections import deque

def shortest_path_bfs_step2_example(graph, start_node, end_node):
    # (Step 1: 초기 설정 부분은 여기에 있다고 가정)
    if start_node == end_node: return 0
    if start_node not in graph: return -1
    queue = deque([(start_node, 0)])
    distances = {start_node: 0}
    print(f"초기 큐: {list(queue)}, 초기 distances: {distances}")

    # --- Step 2: BFS 탐색 시작 (큐가 빌 때까지) ---
    print("\n--- Step 2: BFS 탐색 시작 ---")
    while queue:
        # --- 노드 꺼내기 ---
        current_node, current_distance = queue.popleft()
        print(f"Dequeue: ({current_node}, {current_distance})")

        # --- 목표 도착 확인 ---
        if current_node == end_node:
            print(f"!!! 목표 노드 {end_node} 도달! 최단 거리: {current_distance}!!!")
            return current_distance # 목표 발견 시 즉시 거리 반환

        # (Step 3: 이웃 노드 확인 및 큐 추가 로직은 여기에 온다고 가정)
        print(f"  -> 목표({end_node}) 아님. 이웃 탐색 진행...")
        # 이 예시에서는 이웃 탐색 없이 루프를 몇 번만 돌고 종료합니다.
        if current_distance >= 1: # 예시를 위한 강제 종료 조건
             print("  (예시 종료: 이웃 탐색 생략)")
             break
        # 실제 코드에서는 이웃 탐색 후 큐에 추가하는 로직이 필요합니다.
        # 예시를 위해 가짜 이웃을 큐에 추가해봅니다.
        if current_node == 0 and 1 in graph.get(0,):
             if 1 not in distances:
                 distances = current_distance + 1
                 queue.append((1, current_distance + 1))
                 print(f"  (예시: 이웃 1 큐에 추가)")


    print("--- 탐색 종료 (목표 미발견 또는 예시 종료) ---")
    return -1 # 목표를 못 찾고 큐가 비면 -1 반환

# 예제 그래프 (테스트용)
graph_ex = {
    #  0: , 1: , 2: , 3:
}
print("--- 테스트 1: 0 -> 3 ---")
shortest_path_bfs_step2_example(graph_ex, 0, 3)
print("\n--- 테스트 2: 0 -> 1 ---")
shortest_path_bfs_step2_example(graph_ex, 0, 1) # 목표를 더 빨리 찾는 경우
````

 ### Step 3: 이웃 노드 확인 및 큐에 추가



 큐에서 꺼낸 노드가 목표 노드가 아니라면, 그 노드에서 갈 수 있는 다음 노드들, 즉 **이웃 노드**들을 탐색해야 합니다.



 1.  **이웃 순회**: `현재 노드`의 모든 `이웃 노드`를 하나씩 확인합니다. 그래프 표현(예: 인접 리스트)에서 현재 노드에 연결된 노드들을 가져옵니다. `graph.get(current_node,)`와 같이 사용하면 `current_node`가 그래프에 키로 존재하지 않는 경우(이론상 발생하기 어려우나 방어적 코딩)에도 에러 없이 빈 리스트를 반환합니다.

 2.  **첫 방문 확인**: 각 `이웃 노드`가 **아직 `distances` 딕셔너리에 없는지** 확인합니다 (`neighbor not in distances`).

     * 딕셔너리에 없다는 것은 해당 노드를 **이번 탐색 경로에서 처음 발견**했다는 의미입니다.

     * BFS는 시작 노드로부터 거리가 가까운 순서대로 탐색하므로, 어떤 노드를 처음 발견했을 때의 경로가 항상 그 노드까지의 최단 경로임을 보장합니다.

 3.  **거리 기록 및 큐 추가**:

     * 만약 `이웃 노드`가 첫 방문이라면 (즉, `distances` 딕셔너리에 없다면):

         * `distances[이웃 노드] = 현재 거리 + 1` 로 시작 노드로부터 해당 이웃 노드까지의 최단 거리를 `distances` 딕셔너리에 기록합니다. 이 기록 자체가 **방문 표시**의 역할도 수행합니다.

         * 다음 탐색을 위해 `(이웃 노드, 현재 거리 + 1)` 튜플을 **큐에 추가**(`queue.append()`)합니다. 이 노드들은 다음 레벨에서 탐색될 것입니다.

 #### 예시코드 3: 이웃 처리

 아래 코드는 `while` 루프 내에서 현재 노드의 이웃들을 확인하고, 아직 방문하지 않은 이웃을 `distances`에 기록하고 큐에 추가하는 과정을 보여줍니다.

````python
# %%
# 예시 코드 3: 이웃 처리 부분
from collections import deque

def shortest_path_bfs_step3_example(graph, start_node, end_node):
    # (Step 1, 2의 일부 코드가 여기에 있다고 가정)
    if start_node == end_node: return 0
    if start_node not in graph: return -1
    queue = deque([(start_node, 0)])
    distances = {start_node: 0}
    print(f"초기 큐: {list(queue)}, 초기 distances: {distances}")

    while queue:
        current_node, current_distance = queue.popleft()
        print(f"\nDequeue: ({current_node}, {current_distance})")

        if current_node == end_node:
            print(f"!!! 목표 노드 {end_node} 도달! 최단 거리: {current_distance}!!!")
            return current_distance

        # --- Step 3: 이웃 노드 확인 및 큐에 추가 ---
        print(f"  -> {current_node}의 이웃 확인:", end=" ")
        # graph.get(current_node,) 로 안전하게 이웃 리스트 가져오기
        for neighbor in graph.get(current_node,):
            print(f" [{neighbor}]", end="")
            # 아직 방문하지 않은 이웃인지 확인 (distances 딕셔너리에 없는지 체크)
            if neighbor not in distances:
                # 첫 방문! 거리 기록 및 방문 처리
                new_distance = current_distance + 1
                distances[neighbor] = new_distance
                # 다음 탐색을 위해 큐에 추가
                queue.append((neighbor, new_distance))
                print(f"(-> 첫 방문! 거리:{new_distance} 기록, 큐 추가)", end="")
            else:
                # 이미 방문한 노드 (더 짧은 경로로 이미 발견됨)
                print(f"(-> 이미 방문)", end="")
        print() # 줄 바꿈
        print(f"  -> 이웃 확인 후 큐 상태: {list(queue)}")
        print(f"  -> 현재 distances 상태: {distances}")

    print("--- 탐색 종료 (목표 미발견) ---")
    return -1

# 예제 그래프 (테스트용)
graph_ex = {
    # 0: ,
    # 1: ,
    # 2: ,
    # 3: ,
    # 4: ,
    # 5:
}
print("--- 테스트: 0 -> 5 ---")
shortest_path_bfs_step3_example(graph_ex, 0, 5)
````

 ### Step 4: 탐색 종료 (목표 노드 못 찾음)



 `while queue:` 루프가 모두 끝났다는 것은 큐가 비었다는 의미입니다. 이는 시작 노드에서 출발하여 도달할 수 있는 모든 노드를 방문했지만, 목표 노드(`end_node`)를 찾지 못했다는 것을 나타냅니다.



 이 경우, 시작 노드에서 목표 노드까지 도달할 수 있는 경로가 존재하지 않는 것입니다. 문제의 요구사항에 따라, 도달할 수 없는 경우 -1을 반환하기로 했으므로, `while` 루프 바깥에서 -1을 반환합니다.

 #### 예시코드 4: 탐색 실패 처리

 아래 코드는 `while` 루프가 종료된 후, 목표 노드를 찾지 못했을 때 -1을 반환하는 부분을 보여줍니다.

````python
# %%
# 예시 코드 4: 탐색 실패 처리 부분
from collections import deque

def shortest_path_bfs_step4_example(graph, start_node, end_node):
    # (Step 1, 2, 3의 코드가 여기에 있다고 가정)
    if start_node == end_node: return 0
    if start_node not in graph: return -1
    queue = deque([(start_node, 0)])
    distances = {start_node: 0}

    while queue:
        current_node, current_distance = queue.popleft()

        if current_node == end_node:
            # print(f"목표 노드 {end_node} 도달! 최단 거리: {current_distance}")
            return current_distance

        for neighbor in graph.get(current_node,):
            if neighbor not in distances:
                distances[neighbor] = current_distance + 1
                queue.append((neighbor, current_distance + 1))

    # --- Step 4: 탐색 종료 (목표 노드 못 찾음) ---
    # while 루프가 끝났다는 것은 큐가 비었다는 의미.
    # 여기까지 왔다면 목표 노드를 찾지 못한 것.
    print(f"--- 탐색 종료: 큐가 비었지만 목표 노드 {end_node}를 찾지 못함 ---")
    return -1 # 큐가 비었는데 목표를 못 찾으면 -1 반환

# 예제 그래프 (연결되지 않은 노드 포함)
graph_unconnected = {
    # 0: ,
    # 1: ,
    # 2: ,
    # 3: ,
    # 4: , # 4, 5는 0, 1, 2, 3과 연결되지 않음
    # 5:
}
print("--- 테스트 1: 0 -> 3 (연결됨) ---")
result1 = shortest_path_bfs_step4_example(graph_unconnected, 0, 3)
print(f"결과: {result1}")

print("\n--- 테스트 2: 0 -> 4 (연결 안됨) ---")
result2 = shortest_path_bfs_step4_example(graph_unconnected, 0, 4)
print(f"결과: {result2}")
````

 ### 💻 전체 코드 구현



 위의 단계들을 종합하여 최단 거리를 찾는 BFS 함수 코드를 완성해 봅시다. 디버깅 및 이해를 돕기 위한 print 문이 포함되어 있습니다.

````python
# %%
# 실습: 특정 노드까지의 최단 거리 구하기 (BFS 활용) - 최종 완성 코드
from collections import deque

def shortest_path_bfs(graph, start_node, end_node):
    """
    가중치 없는 그래프에서 BFS를 사용하여 start_node부터 end_node까지의
    최단 거리를 구하는 함수. 도달 불가 시 -1 반환.
    """
    # --- Step 1: 필요한 것들 준비하기 ---
    if start_node == end_node: # 시작과 목표가 같으면 거리는 0
        print(f"시작 노드와 목표 노드가 {start_node}로 동일합니다.")
        return 0
    # 시작 노드가 그래프에 없는 경우도 확인 (end_node는 없어도 탐색은 가능)
    if start_node not in graph:
        print(f"오류: 시작 노드 {start_node}가 그래프에 없습니다.")
        return -1
    # 목표 노드가 그래프에 없는 경우, 도달 불가능하므로 미리 처리할 수도 있으나,
    # BFS 탐색 과정에서 자연스럽게 처리되므로 필수는 아님.
    # if end_node not in graph:
    #     print(f"정보: 목표 노드 {end_node}가 그래프에 없어 도달 불가능합니다.")
    #     return -1

    # 1. 큐 생성 및 시작점 추가: (노드, 거리)
    queue = deque([(start_node, 0)])
    # 2. 방문 기록 및 거리 저장용 딕셔너리: {노드: 거리}
    distances = {start_node: 0}

    print(f"--- BFS 최단 거리 탐색 시작 ({start_node} -> {end_node}) ---")
    print(f"초기 큐: {list(queue)}, 초기 distances: {distances}")

    # --- Step 2: BFS 탐색 시작 ---
    while queue:
        # --- Step 3 (코드 상): 노드 꺼내기 및 목표 도착 확인 ---
        current_node, current_distance = queue.popleft()
        print(f"\nDequeue: ({current_node}, {current_distance}), 현재 distances: {distances}")

        # 목표 노드 도착 확인
        if current_node == end_node:
            print(f"--- 목표 노드 {end_node} 도달! 최단 거리: {current_distance} ---")
            return current_distance # 목표 발견 시 즉시 거리 반환

        # --- Step 3 (코드 상): 이웃 노드 확인 및 큐에 추가 ---
        print(f"  -> {current_node}의 이웃 확인:", end=" ")
        # graph.get()을 사용하여 키가 없는 경우에도 안전하게 처리
        for neighbor in graph.get(current_node,):
            print(f" [{neighbor}]", end="")
            # 아직 방문하지 않은 이웃인지 확인 (distances 딕셔너리에 없는지 체크)
            if neighbor not in distances:
                # 첫 방문! 거리 기록 및 방문 처리
                new_distance = current_distance + 1
                distances[neighbor] = new_distance
                # 다음 탐색을 위해 큐에 추가
                queue.append((neighbor, new_distance))
                print(f"(-> 첫 방문! 거리:{new_distance} 기록, 큐 추가)", end="")
            else:
                print(f"(-> 이미 방문)", end="")
        print() # 줄 바꿈
        print(f"  -> 이웃 확인 후 큐 상태: {list(queue)}")


    # --- Step 4: 탐색 종료 (목표 노드 못 찾음) ---
    # while 루프가 끝났는데 return되지 않았다면 목표 노드를 못 찾은 것
    print(f"--- 탐색 종료: 목표 노드 {end_node}에 도달할 수 없음 ---")
    return -1 # 큐가 비었는데 목표를 못 찾으면 -1 반환

# 예제 그래프
graph_sp = {
    # 0: ,
    # 1: ,
    # 2: ,
    # 3: ,
    # 4: ,
    # 5: ,
    # 6: , # 6, 7은 다른 노드들과 연결되지 않음
    # 7:   # 7은 아무데도 가지 않음
}

# 테스트 케이스 실행
print(f"\n결과 => 0 -> 5 최단 거리: {shortest_path_bfs(graph_sp, 0, 5)}") # 예상: 3
print("=" * 50)
print(f"\n결과 => 0 -> 0 최단 거리: {shortest_path_bfs(graph_sp, 0, 0)}") # 예상: 0
print("=" * 50)
print(f"\n결과 => 3 -> 2 최단 거리: {shortest_path_bfs(graph_sp, 3, 2)}") # 예상: 3 (3 -> 1 -> 0 -> 2 또는 3 -> 1 -> 4 -> 2 또는 3 -> 5 -> 4 -> 2 등)
print("=" * 50)
print(f"\n결과 => 0 -> 6 최단 거리: {shortest_path_bfs(graph_sp, 0, 6)}") # 예상: -1 (연결 안됨)
print("=" * 50)
print(f"\n결과 => 6 -> 7 최단 거리: {shortest_path_bfs(graph_sp, 6, 7)}") # 예상: 1
print("=" * 50)
print(f"\n결과 => 7 -> 0 최단 거리: {shortest_path_bfs(graph_sp, 7, 0)}") # 예상: -1 (연결 안됨)
print("=" * 50)
print(f"\n결과 => 8 -> 0 최단 거리: {shortest_path_bfs(graph_sp, 8, 0)}") # 예상: -1 (시작 노드 8이 그래프에 없음)
print("=" * 50)
print(f"\n결과 => 0 -> 8 최단 거리: {shortest_path_bfs(graph_sp, 0, 8)}") # 예상: -1 (목표 노드 8에 도달 불가)
print("=" * 50)
````

 * DFS 알고리즘의 **시간 복잡도**와 **공간 복잡도**를 그래프 표현 방식(인접 리스트) 기준으로 분석할 수 있다. ⏱️💾

 * **BFS와 DFS**의 탐색 방식, 특징, 활용 사례의 **차이점**을 명확히 **비교**하고, 문제 상황에 **더 적합한 탐색 알고리즘**을 선택할 수 있다. 🤔💡

 ## 📚 깊이 우선 탐색 (Depth-First Search, DFS) 란? 🌲➡️🔙



 * **정의**: 그래프 탐색 알고리즘 중 하나로, 시작 정점에서 출발하여 **한 방향으로 갈 수 있을 때까지 최대한 깊게** 들어간 후,

 더 이상 갈 곳이 없으면 **막다른 길에서 되돌아 나와(backtrack)** 다른 방향으로 탐색을 계속하는 방법.

 * **핵심 아이디어**: 현재 경로에서 갈 수 있는 가장 깊은 곳까지 우선적으로 탐색하고, 막히면 이전 갈림길로 돌아와 다른 경로를 시도.

 * **주요 특징**:

     * **스택(Stack)** 자료구조 또는 **재귀 함수(Recursion)** 를 사용하여 구현 (LIFO: Last-In, First-Out).

     * 찾은 경로가 **최단 경로임을 보장하지 않음**. 🔴

     * 그래프의 모든 정점을 방문하는 데 사용됨.

 * **비유**:

     * 미로 찾기: 한쪽 길을 선택해서 계속 가다가 막다른 길을 만나면, 이전 갈림길로 돌아와 다른 길을 시도하는 방식. 🗺️

     * 트리(Tree) 순회: 특정 자식 노드의 서브트리를 끝까지 탐색한 후, 다른 자식 노드의 서브트리를 탐색하는 방식 (전위/중위/후위 순회와 유사). 🌲

 * *[시각 자료 삽입: DFS 탐색 과정을 단계별로 보여주는 애니메이션 또는 그림 (스택/재귀 방식)]*

 ## ⚙️ DFS 알고리즘 동작 과정



 DFS는 스택 또는 재귀를 사용하여 구현할 수 있습니다.



 ### 1. 스택(Stack)을 이용한 반복적(Iterative) DFS



 1.  **초기화**:

     * 빈 **스택(Stack)** 을 생성합니다. (파이썬에서는 `list` 또는 `collections.deque` 사용 가능)

     * **방문 여부**를 기록할 자료구조(예: `set`)를 생성합니다.

     * **시작 노드(start_node)** 를 스택에 넣습니다(push).



 2.  **반복 (스택이 빌 때까지)**:

     * **Pop**: 스택에서 노드 하나를 꺼냅니다 (`current_node`).

     * **방문 확인 및 처리**:

         * 만약 `current_node`가 **아직 방문하지 않은 상태**라면:

             * `current_node`를 **방문 표시**합니다.

             * `current_node`에 대해 원하는 **작업을 수행**합니다. (예: 출력)

             * **이웃 Push**: `current_node`의 **모든 이웃(neighbor)** 들을 스택에 넣습니다(push).

                 * (주의: 이웃을 스택에 넣는 순서에 따라 실제 방문 순서가 달라질 수 있음. 보통 역순으로 넣어야 원하는 순서대로 pop됨)



 3.  **종료**: 스택이 비면 탐색을 종료합니다.



 * **왜 스택을 사용할까?**: 스택의 **LIFO(후입선출)** 특징 때문에, 가장 최근에 스택에 추가된 (현재 경로에서 더 깊이 들어간) 노드의 이웃이 먼저 탐색 대상이 됩니다.

 막다른 길에 도달하면 더 이상 이웃이 스택에 추가되지 않고, 스택에서 이전 노드가 pop되어 다른 경로 탐색을 시작하게 됩니다 (백트래킹).



 ### 2. 재귀(Recursion)를 이용한 DFS



 1.  **초기화**:

     * **방문 여부**를 기록할 자료구조(예: `set`)를 생성합니다.

     * **DFS 함수 정의**: `dfs(current_node, visited)`

         * 함수 시작 시, `current_node`를 **방문 표시**합니다.

         * `current_node`에 대해 원하는 **작업을 수행**합니다.

         * `current_node`의 **모든 이웃(neighbor)** 들을 확인합니다.

             * 각 이웃 `neighbor`에 대해:

                 * 만약 `neighbor`가 **아직 방문하지 않은 상태**라면:

                     * `dfs(neighbor, visited)` 를 **재귀 호출**합니다.

 2.  **시작**: `dfs(start_node, visited)` 를 호출하여 탐색을 시작합니다.



 * **왜 재귀를 사용할까?**: 재귀 호출 자체가 **함수 호출 스택(Call Stack)** 을 사용하므로, 스택을 이용한 반복적 DFS와 동일한 방식으로 동작합니다. 함수 호출 시 현재 상태가 스택에 저장되고, 재귀 호출이 깊어지면 스택이 쌓입니다. 재귀 호출이 끝나면(막다른 길 도달) 스택에서 이전 함수 상태로 돌아와(pop) 다른 경로 탐색을 이어갑니다. 코드가 더 간결해 보일 수 있습니다.

 ---



 ## 🤔 DFS (깊이 우선 탐색) 핵심 아이디어: 한 길로 끝까지 가보자!



 DFS는 마치 **미로 찾기**를 할 때 한쪽 벽을 따라 **갈 수 있는 데까지 쭉 가보고**,

 막다른 길에 다다르면 **다시 돌아 나와 다른 길**로 가보는 방식과 비슷합니다.



 1.  **시작점**에서 출발하여 **한 방향**으로 계속 나아갑니다.

 2.  더 이상 갈 곳이 없는 **막다른 길**에 도착하면, **바로 직전 갈림길**로 돌아옵니다(백트래킹).

 3.  돌아온 갈림길에서 **아직 가보지 않은 다른 방향**으로 다시 탐색을 시작합니다.

 4.  모든 길을 다 가볼 때까지 이 과정을 반복합니다.



 이 방식 때문에 DFS는 **그래프의 모든 노드를 방문**하거나,

 **경로의 존재 여부**를 확인할 때 자주 사용됩니다.

 ---



 ## ⚙️ DFS 구현 방법 1: 스택(Stack) 활용하기



 스택은 **나중에 넣은 것을 먼저 꺼내는(LIFO)** 특징이 있습니다.

 DFS에서는 스택을 "다음에 가볼 곳 목록"으로 사용합니다.



 1.  **시작**:

     * 탐색 **시작 노드**를 **스택**에 넣습니다.

     * "방문 기록부"(`visited`)를 준비합니다.



 2.  **탐색 (스택이 빌 때까지 반복)**:

     * 스택의 **가장 위(최근에 넣은) 노드**를 꺼냅니다 (`pop`).

     * 이 노드를 **아직 방문 안 했다면**:

         * "방문했다"고 **기록**하고, **처리**합니다 (예: 방문 순서 리스트에 추가).

         * 현재 노드에 **연결된 이웃 노드**들을 확인합니다.

         * 이웃 노드 중 **아직 방문 안 한 노드**들을 **스택에 넣습니다**.

           (새롭게 가볼 곳 목록에 추가)



 3.  **종료**: 스택이 비면, 연결된 모든 노드를 방문한 것입니다.



 **핵심**: 스택에서 꺼낸 노드의 이웃들을 바로 다음에 탐색할 후보로 스택에 넣기 때문에,

 한 방향으로 깊이 들어가는 탐색이 자연스럽게 이루어집니다.

 ### 🐍💻 스택 기반 DFS 코드 (쉬운 버전)

````python
# %%
from collections import deque # deque를 스택처럼 사용할 수 있어요! (append, pop)

def dfs_stack_simple(graph, start_node):
    """스택을 이용한 DFS 방문 순서를 반환하는 간단한 함수"""
    if start_node not in graph:
        print(f"오류: 시작 노드 '{start_node}'가 그래프에 없습니다.")
        return []

    visited = set()           # 방문했던 노드 기록용 (중복 방지)
    stack = deque([start_node]) # 다음에 갈 노드 목록 (스택 역할)
    visited_order = []      # 최종 방문 순서 기록

    print(f"--- DFS (스택) 탐색 시작 (시작: '{start_node}') ---")

    while stack: # 스택에 갈 곳이 남아 있는 동안
        # 1. 스택에서 가장 최근에 넣은 노드 꺼내기
        current_node = stack.pop()
        print(f"\n스택 Pop: '{current_node}' (남은 스택: {list(stack)})")

        # 2. 이미 방문했던 곳인지 확인
        if current_node not in visited:
            # 3. 처음 방문! -> 방문 기록 및 처리
            print(f"  -> '{current_node}' 첫 방문! 방문 처리합니다.")
            visited.add(current_node)
            visited_order.append(current_node)

            # 4. 현재 노드의 이웃들을 스택에 추가 (다음 갈 곳 후보)
            print(f"  -> '{current_node}'의 이웃 확인:", end="")
            # 참고: reversed() 를 써서 넣으면 보통 재귀 방식과 비슷하게 탐색됩니다.
            #       여기서는 그래프 딕셔너리 순서대로 넣어봅니다.
            for neighbor in reversed(graph.get(current_node, [])):
                if neighbor not in visited:
                    print(f" '{neighbor}'(스택추가)", end="")
                    stack.append(neighbor) # 방문 안 한 이웃만 스택에 push
            print() # 줄 바꿈
        else:
            print(f"  -> '{current_node}'는 이미 방문했습니다.")

    print("\n--- DFS (스택) 탐색 종료 ---")
    return visited_order

# 예제 그래프
graph_example = {
    'A': ['B', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['B'],
    'D': ['A', 'E'],
    'E': ['B', 'D']
}

# DFS (스택) 실행
dfs_stack_result = dfs_stack_simple(graph_example, 'A')
print(f"\n최종 DFS (스택) 방문 순서: {dfs_stack_result}")
# (reversed 사용 시) 출력 예시: ['A', 'B', 'C', 'E', 'D'] - 이웃 추가/pop 순서 따라 다름
````

 ## ⚙️ DFS 구현 방법 2: 재귀(Recursion) 활용하기



 재귀는 함수가 자기 자신을 다시 호출하는 방식입니다. DFS의 "한 길로 끝까지 가보고 돌아와서 다른 길 가기" 특징을 재귀로 자연스럽게 표현할 수 있습니다.



 **동작 방식 (함수 `dfs(현재 노드)`):**



 1.  **도착**: `현재 노드`에 도착하면,

     * "방문했다"고 **기록**하고 **처리**합니다.

 2.  **길 찾기**: `현재 노드`에 연결된 **이웃 노드**들을 살펴봅니다.

 3.  **새 길 탐색**: 만약 이웃 노드가 **아직 방문 안 한 곳**이라면,

     * **그 이웃 노드에서 다시 `dfs()` 함수를 호출**합니다 (새로운 길 탐색 시작!).

 4.  **막다른 길/복귀**: 모든 이웃을 확인했거나, 이미 방문한 이웃뿐이라면

     함수 호출이 끝나고 **이전 갈림길(호출했던 곳)로 돌아갑니다.**



 **핵심**: 함수 호출 자체가 스택처럼 쌓이며 "갔던 길"을 기억해주고,

 함수가 끝나면 자동으로 "돌아오는" 효과가 있어 코드가 간결해집니다.

 ### 🐍💻 재귀 기반 DFS 코드 (쉬운 버전)

````python
# %%
# 재귀 깊이 제한을 늘려둘 수 있습니다 (그래프가 매우 깊을 경우 필요)
# import sys
# sys.setrecursionlimit(10**6)

visited_recursive = set() # 방문 기록용 집합 (재귀 호출 간에 공유되어야 함)
dfs_order_recursive = [] # 방문 순서 기록용 리스트 (재귀 호출 간에 공유)

def dfs_recursive_simple(graph, current_node):
    """재귀를 이용한 DFS 탐색 함수"""
    # 1. 현재 노드 방문 처리
    print(f"  -> 현재 노드: '{current_node}' 방문 처리")
    visited_recursive.add(current_node)
    dfs_order_recursive.append(current_node)

    # 2. 현재 노드의 이웃들을 확인
    print(f"  -> '{current_node}'의 이웃 확인:", end="")
    for neighbor in graph.get(current_node, []):
        print(f" '{neighbor}'", end="")
        # 3. 방문하지 않은 이웃이라면, 재귀 호출!
        if neighbor not in visited_recursive:
            print(f"(-> 재귀호출!)", end="")
            dfs_recursive_simple(graph, neighbor) # 여기서 더 깊이 들어감
        # (이미 방문한 이웃은 그냥 넘어감)
    print(f" ( '{current_node}'의 모든 이웃 확인 완료)") # 함수 종료 및 복귀 지점


# --- 재귀 DFS 실행을 위한 준비 ---
graph_example = {
    'A': ['B', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['B'],
    'D': ['A', 'E'],
    'E': ['B', 'D']
}
start_node_rec = 'A'

# 전역 변수 초기화 (매 실행 시 필요)
visited_recursive.clear()
dfs_order_recursive.clear()

print(f"--- DFS (재귀) 탐색 시작 (시작: '{start_node_rec}') ---")
if start_node_rec in graph_example:
    # 4. 첫 재귀 함수 호출 시작
    dfs_recursive_simple(graph_example, start_node_rec)
else:
    print(f"오류: 시작 노드 '{start_node_rec}'가 그래프에 없습니다.")
print("\n--- DFS (재귀) 탐색 종료 ---")

print(f"\n최종 DFS (재귀) 방문 순서: {dfs_order_recursive}")
# 출력 예시: ['A', 'B', 'C', 'E', 'D'] - 이웃 순회 순서에 따라 달라짐
````

 ## ⏱️💾 DFS 복잡도: BFS와 같아요!



 (V: 노드 개수, E: 간선 개수)



 * **시간 복잡도: O(V + E)** 🚀

   * 모든 노드(V)를 딱 한 번씩 방문하고,

   * 모든 간선(E)을 딱 한 번씩 검사하기 때문입니다. (스택/재귀 방식 동일)



 * **공간 복잡도: O(V)** 💾

   * **스택 방식**: 스택에 최악의 경우 모든 노드(V)가 쌓일 수 있습니다.

   * **재귀 방식**: 함수 호출 스택(Call Stack)이 가장 깊게 들어갔을 때의 깊이가

     최악의 경우 모든 노드(V)만큼 될 수 있습니다.

   * **방문 기록(visited)** 도 모든 노드(V) 정보를 저장해야 합니다.



 **결론**: DFS도 BFS와 마찬가지로 매우 **효율적인** 그래프 탐색 방법입니다!

 ## 🛤️🔗 DFS 활용 사례



 DFS는 특유의 깊이 우선 탐색 방식으로 인해 다음과 같은 문제들에 유용하게 사용됩니다.



 1.  **경로 찾기 (Path Finding)**:

     * 두 노드 사이에 경로가 **존재하는지 여부**를 찾는 데 사용될 수 있습니다. DFS 탐색 중 목표 노드를 만나면 경로가 존재한다고 판단할 수 있습니다. (최단 경로 보장은 안 됨)



 2.  **연결 요소 (Connected Components) 찾기**:

     * BFS와 마찬가지로, 무방향 그래프에서 연결된 노드들의 집합을 찾는 데 사용될 수 있습니다. 방문하지 않은 노드에서 DFS를 시작하여 도달 가능한 모든 노드를 찾으면 하나의 연결 요소가 됩니다.



 3.  **사이클 탐지 (Cycle Detection)**:

     * 그래프 내에 사이클이 존재하는지 확인하는 데 DFS를 응용할 수 있습니다. 탐색 중 현재 경로 상에 이미 방문했던 노드를 다시 만나면 사이클이 존재한다고 판단합니다. (약간의 추가 로직 필요)



 4.  **위상 정렬 (Topological Sort)**:

     * 방향 비순환 그래프(DAG, Directed Acyclic Graph)에서 노드들의 선후 관계를 유지하며 일렬로 정렬하는 위상 정렬에 DFS가 핵심적으로 사용됩니다.



 5.  **백트래킹 (Backtracking) 알고리즘의 기반**:

     * 모든 가능한 해를 탐색하다가 특정 경로가 해답이 될 수 없다고 판단되면 이전 상태로 되돌아가 다른 경로를 탐색하는 백트래킹 알고리즘은 DFS의 아이디어를 기반으로 합니다. (예: 미로 찾기, N-Queens 문제 등) (6교시 내용)

 ---



 ## 🤔💡 BFS vs DFS 비교 및 선택 가이드



 BFS와 DFS는 그래프 탐색의 대표적인 두 알고리즘이지만, 탐색 방식과 특징에 차이가 있어 문제 상황에 맞게 선택해야 합니다.



 | 특징             | BFS (너비 우선 탐색)                     | DFS (깊이 우선 탐색)                       |
 | :--------------- | :--------------------------------------- | :----------------------------------------- |
 | **탐색 방식** | 시작 노드 기준 **가까운(넓은)** 순서 (레벨별) | 한 방향으로 **최대한 깊게** 탐색 후 백트래킹 |
 | **구현 자료구조**| **큐 (Queue)** - FIFO                    | **스택 (Stack)** 또는 **재귀 (Recursion)** - LIFO |
 | **최단 경로** | **보장 (가중치 없는 그래프)** 🥇          | 보장 안 함                                 |
 | **경로 길이** | 일반적으로 더 짧은 경로 먼저 발견        | 경로 길이가 다양하게 나타날 수 있음        |
 | **메모리 사용** | 레벨 폭 넓으면 큐 메모리 많이 사용 가능    | 경로 깊이 깊으면 스택 메모리 많이 사용 가능  |
 | **주요 활용** | 최단 경로, 연결 요소, 네트워크 브로드캐스트 | 경로 존재 확인, 사이클 탐지, 위상 정렬, 백트래킹 |



 **언제 무엇을 쓸까?**



 * **최단 경로(최소 간선 수)를 찾아야 할 때** → **BFS**

 * **경로의 존재 여부만 확인**하면 될 때 → **DFS** (구현이 더 간단할 수 있음)

 * **그래프의 모든 노드를 방문**해야 할 때 → **BFS, DFS 둘 다 가능** (상황에 따라 선택)

 * **탐색 깊이가 매우 깊어질 가능성**이 있고 스택 오버플로우가 우려될 때 → **BFS** 또는 **반복적 DFS**

 * **해답이 시작 노드에서 멀리 떨어져 있을 가능성**이 높을 때 → **DFS** (더 빨리 찾을 수도 있음)

 * **위상 정렬, 사이클 탐지** 등 특정 알고리즘 구현 시 → **DFS**가 더 자연스러운 경우가 많음

 ## ❓ 객관식, 단답형 및 서술형 문제



 * 각 문제의 난이도를 ⭐️ ~ ⭐️⭐️⭐️⭐️⭐️ 로 표시했습니다.

 ### 🧐 객관식 문제 1 (난이도: ⭐️⭐️)



 깊이 우선 탐색(DFS) 알고리즘에 대한 설명으로 **가장 적절한** 것은 무엇인가요?



 1.  큐(Queue)를 사용하여 구현하며, 레벨 순서로 탐색한다.

 2.  시작 노드에서 가장 가까운 노드들을 먼저 방문한다.

 3.  가중치 없는 그래프에서 항상 최단 경로를 찾는다.

 4.  스택(Stack) 또는 재귀(Recursion)를 사용하여 한 경로를 깊게 탐색한다.

 5.  시간 복잡도는 항상 O(V²)이다.

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 DFS(깊이 우선 탐색)는 한 방향으로 가능한 깊이 탐색하다가 막다른 길에 도달하면 되돌아 나와 다른 경로를 탐색하는 방식입니다. 이러한 LIFO(후입선출) 방식의 탐색은 스택 자료구조나 재귀 함수의 호출 스택을 이용하여 자연스럽게 구현할 수 있습니다.

 1. 큐와 레벨 순서는 BFS의 특징입니다.

 2. 가까운 노드를 먼저 방문하는 것은 BFS의 특징입니다.

 3. 최단 경로 보장은 BFS의 특징입니다 (가중치 없는 그래프).

 5. 인접 리스트 사용 시 시간 복잡도는 O(V + E)입니다.

 ### 🧐 객관식 문제 2 (난이도: ⭐️⭐️⭐️)



 다음 중 DFS 알고리즘이 BFS보다 **더 적합**하다고 일반적으로 여겨지는 문제 유형은 무엇인가요?



 1.  가중치가 없는 그래프에서 두 노드 간 최단 거리(간선 수) 찾기

 2.  소셜 네트워크에서 특정 인물로부터 k 단계 이내의 모든 친구 찾기

 3.  미로 찾기 게임에서 출발점에서 도착점까지 가는 경로 존재 여부 확인하기

 4.  네트워크에서 특정 서버로부터 연결된 모든 서버 목록 찾기 (최단 거리 무관)

 5.  두 단어 간의 최소 변환 횟수 찾기 (예: hit -> cog)

 ### 정답 및 해설

 **정답: 3번**



 **해설:**

 DFS는 경로의 존재 여부를 확인하거나, 특정 조건을 만족하는 깊은 경로를 찾는 데 강점이 있습니다. 미로 찾기에서 도착점까지 가는 경로가 있는지 없는지만 확인하는 경우, DFS는 한 경로를 깊게 탐색하므로 BFS보다 더 빨리 답을 찾을 수도 있습니다 (최단 경로가 아니라면).

 1. 최단 거리는 BFS가 보장합니다.

 2. k 단계 이내의 탐색은 레벨별 탐색인 BFS가 더 자연스럽습니다.

 4. 연결된 모든 서버 목록은 BFS, DFS 둘 다 가능하지만, 특별히 깊이 탐색할 이유가 없다면 BFS도 좋습니다.

 5. 최소 변환 횟수는 최단 경로 문제와 유사하므로 BFS가 더 적합합니다.



 (참고: 3번 문제도 BFS로 풀 수 있으며, 최단 경로를 찾으려면 BFS를 사용해야 합니다. 하지만 단순히 경로 존재 여부만 묻는다면 DFS도 효과적인 선택지가 될 수 있습니다.)

 ### 🧐 객관식 문제 3 (난이도: ⭐️⭐️⭐️⭐️)



 다음 그래프에서 노드 'A'를 시작으로 **스택(Stack) 기반 DFS**를 수행할 때,

 방문 순서로 **가능하지 않은** 것은 무엇인가요?

 (단, 스택에 이웃 노드를 넣을 때는 **알파벳 역순**으로 넣는다고 가정합니다.

 이렇게 하면 스택에서 꺼낼(pop) 때는 알파벳 오름차순으로 처리됩니다.)



 ```

   A --- B --- C

   |     |

   D --- E

 ```

 (간선: A-B, A-D, B-C, B-E, D-E)



 **그래프 (인접 리스트):**

 ```python

 graph = {

     'A': ['B', 'D'],

     'B': ['A', 'C', 'E'],

     'C': ['B'],

     'D': ['A', 'E'],

     'E': ['B', 'D']

 }

 ```



 1. A, B, C, E, D

 2. A, D, E, B, C

 3. A, B, E, D, C

 4. A, B, E, C, D

 5. A, C, B, E, D

 ### 정답 및 해설

 **정답: 5번**



 **해설:**

 스택 기반 DFS는 LIFO(후입선출) 원칙을 따릅니다. 문제의 조건에 따라,

 현재 노드에서 방문하지 않은 이웃 노드들을 스택에 추가할 때 **알파벳 역순**으로

 넣습니다. 이는 스택에서 꺼낼(pop) 때 **알파벳 오름차순**으로 꺼내서

 탐색을 진행하게 만듭니다.



 **단계별 탐색 과정:**



 1.  **시작**:

     * `stack = deque(['A'])`

     * `visited = set()`

     * `방문 순서 = []`



 2.  **Pop 'A'**:

     * 'A' 방문 안 함 -> `visited = {'A'}`, `방문 순서 = ['A']`

     * 'A'의 이웃: 'B', 'D'. 알파벳 역순('D', 'B')으로 스택에 push.

     * `stack = deque(['D', 'B'])`



 3.  **Pop 'B'**:

     * 'B' 방문 안 함 -> `visited = {'A', 'B'}`, `방문 순서 = ['A', 'B']`

     * 'B'의 이웃: 'A', 'C', 'E'. 방문 안 한 이웃은 'C', 'E'.

     * 알파벳 역순('E', 'C')으로 스택에 push.

     * `stack = deque(['D', 'E', 'C'])`



 4.  **Pop 'C'**:

     * 'C' 방문 안 함 -> `visited = {'A', 'B', 'C'}`, `방문 순서 = ['A', 'B', 'C']`

     * 'C'의 이웃: 'B'. 이미 방문함. (스택 추가 없음)

     * `stack = deque(['D', 'E'])`



 5.  **Pop 'E'**:

     * 'E' 방문 안 함 -> `visited = {'A', 'B', 'C', 'E'}`, `방문 순서 = ['A', 'B', 'C', 'E']`

     * 'E'의 이웃: 'B', 'D'. 방문 안 한 이웃은 'D'.

     * 알파벳 역순 (어차피 하나) 'D'를 스택에 push.

     * `stack = deque(['D', 'D'])`



 6.  **Pop 'D'**:

     * 'D' 방문 안 함 -> `visited = {'A', 'B', 'C', 'E', 'D'}`, `방문 순서 = ['A', 'B', 'C', 'E', 'D']`

     * 'D'의 이웃: 'A', 'E'. 둘 다 이미 방문함. (스택 추가 없음)

     * `stack = deque(['D'])`



 7.  **Pop 'D'**:

     * 'D'는 이미 방문함. 무시.

     * `stack = deque([])`



 8.  **종료**: 스택이 비었으므로 종료.



 **결과:** 주어진 규칙에 따른 유일한 방문 순서는 **A, B, C, E, D** 입니다. (선택지 1번)



 **가능하지 않은 순서 찾기:**

 문제에서 주어진 규칙을 따르면 방문 순서는 `A, B, C, E, D` 하나로 결정됩니다.

 따라서 이 순서가 아닌 다른 모든 선택지는 **가능하지 않습니다.**



 * **선택지 5번 (`A, C, B, E, D`)** 는 특히 불가능합니다. 시작 노드 'A' 다음에는 이웃 중 알파벳 순서가 가장 빠른 'B'가 pop되어 방문되어야 하는데, 'C'가 먼저 방문되었으므로 규칙에 어긋납니다. 다른 선택지들(2, 3, 4)도 모두 위에서 유도된 유일한 순서와 다르므로 가능하지 않습니다. (여기서는 5번을 대표적인 오답으로 선택)

 ### ✍️ 단답형 / 서술형 문제 1 (난이도: ⭐️⭐️⭐️)



 DFS(깊이 우선 탐색)와 BFS(너비 우선 탐색)는 그래프 탐색의 대표적인 두 가지 방법입니다. 두 알고리즘의 **탐색 순서**와 사용하는 **주요 자료구조**의 차이점을 설명하세요.

 ### 정답 및 해설

 **정답:**



 **탐색 순서:**

 * **BFS (너비 우선 탐색)**: 시작 노드로부터 **가까운 거리(레벨) 순서대로** 탐색합니다. 즉, 시작 노드의 모든 이웃(레벨 1)을 먼저 방문하고, 그 다음 레벨 1 노드들의 모든 이웃(레벨 2)을 방문하는 방식으로, **넓게 퍼져나가며** 탐색합니다.

 * **DFS (깊이 우선 탐색)**: 시작 노드에서 출발하여 **한 방향으로 최대한 깊이** 들어갑니다. 막다른 길에 도달하면 **되돌아와서(backtrack)** 아직 방문하지 않은 다른 경로를 따라 다시 깊이 탐색합니다. 즉, **하나의 경로를 끝까지 파고드는** 방식으로 탐색합니다.



 **주요 자료구조:**

 * **BFS**: **큐(Queue)** 를 사용합니다. 큐의 FIFO(선입선출) 특성을 이용하여 먼저 발견된(가까운) 노드를 먼저 처리함으로써 레벨별 탐색 순서를 유지합니다.

 * **DFS**: **스택(Stack)** 또는 **재귀 호출(Recursion)** 을 사용합니다. 스택의 LIFO(후입선출) 특성 또는 재귀 함수의 호출 스택을 이용하여 가장 최근에 방문한 노드(더 깊이 들어간 노드)의 이웃을 먼저 탐색하고, 막다른 길에서 되돌아오는(백트래킹) 동작을 구현합니다.

 ### ✍️ 단답형 / 서술형 문제 2 (난이도: ⭐️⭐️⭐️⭐️)



 재귀 함수를 사용하여 DFS를 구현할 때 발생할 수 있는 **잠재적인 문제점**은 무엇이며, 이 문제점을 **완화**할 수 있는 방법은 무엇인가요?

 ### 정답 및 해설

 **정답:**



 **잠재적인 문제점:**

 재귀 함수를 사용하여 DFS를 구현할 때 발생할 수 있는 가장 큰 잠재적인 문제점은 **스택 오버플로우(Stack Overflow) 에러**입니다. 재귀 함수는 호출될 때마다 함수 호출 정보(매개변수, 지역 변수, 복귀 주소 등)를 **호출 스택(Call Stack)** 메모리 영역에 쌓습니다. 만약 그래프의 깊이가 매우 깊거나 특정 경로가 매우 길 경우, 재귀 호출이 너무 많이 중첩되어 시스템이 할당한 호출 스택 메모리 용량을 초과하게 되고, 이로 인해 스택 오버플로우 에러가 발생하며 프로그램이 비정상적으로 종료될 수 있습니다.



 **완화 방법:**



 1.  **반복적(Iterative) DFS 사용**: 재귀 호출 대신 **명시적인 스택(Stack) 자료구조**를 사용하여 DFS를 구현하는 것입니다. 이 방법은 시스템 호출 스택의 제한에 영향을 받지 않으므로, 그래프의 깊이가 매우 깊더라도 스택 오버플로우 없이 탐색을 수행할 수 있습니다. (메모리가 허용하는 한)

 2.  **재귀 깊이 제한 늘리기 (주의해서 사용)**: 파이썬의 경우 `sys` 모듈을 사용하여 재귀 호출의 최대 깊이 제한을 늘릴 수 있습니다 (`import sys; sys.setrecursionlimit(새로운_한계값)`). 하지만 이는 근본적인 해결책이 아니며, 시스템의 실제 스택 메모리 크기를 초과하면 여전히 문제가 발생할 수 있고, 너무 큰 값으로 설정하면 다른 예기치 못한 문제를 일으킬 수 있으므로 신중하게 사용해야 합니다. 일반적으로는 반복적 DFS 구현을 더 권장합니다.

 * 재귀 기반 DFS 함수 (`dfs_recursive_helper`) 구조를 활용합니다.

 * 함수 시작 시 현재 노드를 방문 처리합니다.

 * 현재 노드가 `end_node`인지 확인합니다. 맞다면 `True`를 반환합니다. (주의: 재귀 호출 결과를 잘 처리해야 함)

 * 방문하지 않은 이웃에 대해 재귀 호출을 수행합니다.

 * 만약 **재귀 호출의 결과가 `True`** 라면, 경로를 찾았다는 의미이므로 **즉시 `True`를 반환**하여 상위 호출로 전파합니다.

 * 모든 이웃에 대한 재귀 호출이 끝났는데도 `True`가 반환되지 않았다면, 현재 경로에서는 `end_node`를 찾지 못한 것이므로 `False`를 반환합니다.

 * 초기 호출 함수(`dfs_recursive`)에서는 `visited` 세트를 만들고 도우미 함수를 호출한 뒤 그 결과를 반환합니다.

### 정답 코드

````python
# %%
# 실습 1: 경로 존재 확인 (DFS 활용)
from collections import deque

# --- 스택 기반 DFS ---
def path_exists_dfs_iterative(graph, start_node, end_node):
    """스택 기반 DFS로 경로 존재 여부 확인"""
    if start_node == end_node: return True
    if start_node not in graph: return False

    visited = set()
    stack = deque([start_node])

    while stack:
        current_node = stack.pop()

        if current_node == end_node: # 목표 노드 발견!
            return True

        if current_node not in visited:
            visited.add(current_node)
            # print(f"Iterative Visiting: {current_node}") # 방문 순서 확인용

            if current_node in graph:
                # 이웃을 역순으로 넣어야 재귀와 유사한 순서 가능성 높음
                for neighbor in reversed(graph[current_node]):
                    if neighbor not in visited:
                        stack.append(neighbor)

    return False # 큐가 비었는데 목표 노드 못 찾음

# --- 재귀 기반 DFS ---
def path_exists_dfs_recursive_helper(graph, current_node, end_node, visited):
    """재귀 DFS 도우미 함수"""
    visited.add(current_node)
    # print(f"Recursive Visiting: {current_node}") # 방문 순서 확인용

    if current_node == end_node: # 목표 노드 발견!
        return True

    if current_node in graph:
        for neighbor in graph[current_node]:
            if neighbor not in visited:
                # 재귀 호출 결과가 True이면 경로 찾음 -> 즉시 True 반환
                if path_exists_dfs_recursive_helper(graph, neighbor, end_node, visited):
                    return True

    # 현재 경로에서 목표 노드 못 찾음
    return False

def path_exists_dfs_recursive(graph, start_node, end_node):
    """재귀 기반 DFS로 경로 존재 여부 확인 (초기 호출 함수)"""
    if start_node == end_node: return True
    if start_node not in graph: return False

    visited = set()
    return path_exists_dfs_recursive_helper(graph, start_node, end_node, visited)


# 예제 그래프
graph_path = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 4],
    3: [1, 5],
    4: [1, 2, 5],
    5: [3, 4],
    6: [7] # 연결되지 않은 컴포넌트
}

# 테스트
print("--- 스택 기반 DFS 경로 확인 ---")
print(f"0 -> 5 경로 존재? {path_exists_dfs_iterative(graph_path, 0, 5)}") # True
print(f"0 -> 6 경로 존재? {path_exists_dfs_iterative(graph_path, 0, 6)}") # False
print(f"5 -> 0 경로 존재? {path_exists_dfs_iterative(graph_path, 5, 0)}") # True
print(f"7 -> 6 경로 존재? {path_exists_dfs_iterative(graph_path, 7, 6)}") # False (7은 그래프에 있지만 6으로 가는 경로 없음)

print("\n--- 재귀 기반 DFS 경로 확인 ---")
print(f"0 -> 5 경로 존재? {path_exists_dfs_recursive(graph_path, 0, 5)}") # True
print(f"0 -> 6 경로 존재? {path_exists_dfs_recursive(graph_path, 0, 6)}") # False
print(f"5 -> 0 경로 존재? {path_exists_dfs_recursive(graph_path, 5, 0)}") # True
print(f"7 -> 6 경로 존재? {path_exists_dfs_recursive(graph_path, 7, 6)}") # False
````

 ### 1. 메모이제이션 (Memoization) - 하향식 (Top-Down)



 * **컨셉**: "위에서부터 일단 **재귀**로 풀어! 단, **계산 결과는 꼭 메모**해두자!"

 * **방법**:

     1.  큰 문제(`fib(n)`)에서 시작해서 필요한 작은 문제(`fib(n-1)`, `fib(n-2)`)를 **재귀 호출**해요.

     2.  함수 맨 처음에 **메모장(딕셔너리 등)** 에 `n`에 대한 답이 **이미 적혀 있는지 확인**해요.

     3.  **적혀있으면?** -> 바로 그 값을 사용! (재계산 안 함! 👍)

     4.  **안 적혀있으면?** -> 재귀 호출로 답을 계산하고, 그 결과를 **메모장에 꼭 적어둔 뒤** 돌려줘요.

 * **특징**: 코드가 재귀적이라 문제 정의와 비슷해서 **이해하기 쉬울 수 있어요**. 필요한 답만 계산해요. 하지만 재귀가 너무 깊어지면 에러가 날 수도 있어요.



 * **예시: 피보나치 수열 (메모이제이션)**

````python
# %%
# 메모장 역할을 할 딕셔너리 (계산 결과 저장용)
memo_fib_cache = {}

def fib_memo(n):
    """피보나치 수열 (메모이제이션 - 재귀 + 기록)"""
    # 0 또는 1이면 바로 반환 (가장 작은 문제)
    if n <= 1:
        return n

    # 메모장에 이미 계산된 결과가 있는지 확인
    if n in memo_fib_cache:
        # print(f"  -> fib({n})는 메모장에서 가져옴!") # 확인용 출력
        return memo_fib_cache[n] # 저장된 값 사용!

    # 메모장에 없으면, 재귀로 계산하고 결과를 메모장에 저장
    # print(f"  -> fib({n}) 계산 시작...") # 확인용 출력
    result = fib_memo(n - 1) + fib_memo(n - 2)
    memo_fib_cache[n] = result # 계산 결과 저장!
    return result

# 테스트
n_fib_test = 10
memo_fib_cache = {} # 테스트 전 메모장 비우기
print(f"--- 피보나치({n_fib_test}) 계산 (메모이제이션) ---")
result_memo = fib_memo(n_fib_test)
print(f"결과: {result_memo}")
# print(f"메모장 내용: {memo_fib_cache}") # 저장된 값들 확인
````

 ### 2. 타뷸레이션 (Tabulation) - 상향식 (Bottom-Up)



 * **컨셉**: "가장 쉬운 문제부터 차근차근 풀어서 **표(테이블)를 완성**하자!"

 * **방법**:

     1.  결과를 저장할 **표(리스트/배열)** 를 만들어요 (보통 문제 크기만큼).

     2.  표의 **가장 앞부분(가장 작은 문제)** 의 답을 먼저 채워요. (예: `fib(0)=0`, `fib(1)=1`)

     3.  **반복문**을 이용해서 표의 앞쪽 값들을 참고해 **다음 칸들의 답**을 계산하고 채워나가요. (예: `fib(2) = fib(1) + fib(0)`)

     4.  반복문이 끝나면 표에 모든 답이 순서대로 채워져 있고, 원하는 답을 꺼내면 돼요!

 * **특징**: 반복문이라 **보통 더 빠르고** 재귀 에러 걱정이 없어요. 하지만 코드가 재귀보다 덜 직관적일 수 있어요.



 * **예시: 피보나치 수열 (타뷸레이션)**

````python
# %%
def fib_tab(n):
    """피보나치 수열 (타뷸레이션 - 반복문 + 표)"""
    if n <= 1:
        return n

    # n+1 크기의 표(리스트) 준비 (0번 ~ n번 인덱스 사용)
    dp_table = [0] * (n + 1)

    # 가장 작은 문제의 답 채우기 (기초 값)
    dp_table[0] = 0
    dp_table[1] = 1
    # print(f"초기 테이블: {dp_table}") # 확인용

    # 표 채우기: 2번 칸부터 n번 칸까지
    for i in range(2, n + 1):
        # 바로 앞 두 칸의 값을 더해서 현재 칸 계산
        dp_table[i] = dp_table[i - 1] + dp_table[i - 2]
        # print(f"테이블[{i}] 채우기: {dp_table[i]}") # 확인용

    # n번 칸의 값이 최종 답
    return dp_table[n]

# 테스트
n_fib_test = 10
print(f"--- 피보나치({n_fib_test}) 계산 (타뷸레이션) ---")
result_tab = fib_tab(n_fib_test)
print(f"결과: {result_tab}")
````

 ## 💪 DP 기초 문제: 계단 오르기



 * **문제**: 계단 N칸을 오를 때, 한 번에 1칸 또는 2칸씩만 갈 수 있다면, 꼭대기(N번째 칸)까지 가는 방법은 총 몇 가지일까요?

 * **DP로 생각하기**:

     * N번째 칸에 도착하는 방법은?

         * (N-1)번째 칸에서 1칸 폴짝! 뛰는 방법들의 수

         * (N-2)번째 칸에서 2칸 폴짝! 뛰는 방법들의 수

     * 따라서, `(N칸까지 가는 방법 수) = (N-1칸까지 방법 수) + (N-2칸까지 방법 수)`

     * 이것도 피보나치처럼 **이전 결과들을 이용**하네요! DP 적용 가능!

 * **점화식**: `dp[n] = dp[n-1] + dp[n-2]`

 * **기초 값**: `dp[1] = 1` (1칸 가는 법: '1'), `dp[2] = 2` (2칸 가는 법: '1+1', '2')



 * **구현 (타뷸레이션)**: 피보나치랑 거의 똑같아요!

````python
# %%
def count_stair_ways_tab(n):
    """계단 오르는 방법 수 (타뷸레이션 DP)"""
    if n <= 0: return 0
    if n == 1: return 1
    if n == 2: return 2 # 피보나치와 기초값만 살짝 다름

    # n+1 크기 테이블 (0번 인덱스는 안 쓸 수도 있지만 편의상)
    dp = [0] * (n + 1)
    # 기초 값 설정
    dp[1] = 1
    dp[2] = 2

    # 3번째 칸부터 n번째 칸까지 계산
    for i in range(3, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2] # 점화식은 동일!

    return dp[n]

# 테스트
num_stairs = 5
print(f"--- 계단 오르기 ({num_stairs}칸) ---")
ways = count_stair_ways_tab(num_stairs)
print(f"방법의 수: {ways}") # 예상: 8
````

 동적 계획법(Dynamic Programming)을 적용하기 위한 두 가지 필수 조건은 무엇인가요?



 1.  선형 시간 복잡도, 상수 공간 복잡도

 4.  분할 정복 (Divide and Conquer), 안정 정렬 (Stable Sort)

 5.  상태 공간 트리 (State Space Tree), 가지치기 (Pruning)

 ### 정답 및 해설

 **정답: 3번**



 **해설:**

 동적 계획법(DP)은 주어진 문제를 더 작은 부분 문제로 나누어 해결하는 방식인데, 이때 다음 두 가지 속성이 만족되어야 효율적으로 적용할 수 있습니다.

 * **겹치는 부분 문제 (Overlapping Subproblems)**: 작은 부분 문제들이 반복적으로 나타나야 합니다. DP는 이 중복 계산을 피하기 위해 부분 문제의 해를 저장하고 재활용합니다.

 * **최적 부분 구조 (Optimal Substructure)**: 전체 문제의 최적 해답이 부분 문제들의 최적 해답을 조합하여 구해질 수 있어야 합니다.

 ### 🧐 객관식 문제 2 (난이도: ⭐️⭐️⭐️)



 동적 계획법의 구현 방식 중 **메모이제이션(Memoization)** 에 대한 설명으로 **옳은** 것은 무엇인가요?



 1.  반복문을 사용하여 가장 작은 부분 문제부터 해를 구해 테이블을 채워나간다. (Bottom-Up)

 2.  재귀 함수를 기반으로 하되, 계산된 부분 문제의 결과를 저장하여 재활용한다. (Top-Down)

 3.  항상 타뷸레이션 방식보다 시간 복잡도 측면에서 유리하다.

 4.  재귀 호출을 사용하지 않으므로 스택 오버플로우 위험이 없다.

 5.  모든 부분 문제를 반드시 계산해야 한다.

 ### 정답 및 해설

 **정답: 2번**



 **해설:**

 메모이제이션(Memoization)은 **Top-Down** 방식으로, 큰 문제를 해결하기 위해 **재귀 함수**를 호출하는 구조를 기본으로 합니다. 이때 각 부분 문제의 결과를 계산하면 메모리(캐시)에 저장해두고, 나중에 동일한 부분 문제가 다시 호출되면 저장된 값을 즉시 반환하여 중복 계산을 피합니다.

 1. 타뷸레이션(Tabulation) 방식에 대한 설명입니다.

 3. 일반적으로 타뷸레이션이 재귀 오버헤드가 없어 더 빠를 수 있습니다.

 4. 재귀 호출을 사용하므로 스택 오버플로우 위험이 있습니다.

 5. 실제로 필요한 부분 문제만 계산하게 될 수 있습니다 (타뷸레이션은 보통 모든 관련 부분 문제를 계산).

 ### ✍️ 단답형 / 서술형 문제 1 (난이도: ⭐️⭐️⭐️⭐️)



 동적 계획법(DP)의 두 가지 구현 방식인 **메모이제이션(Memoization)** 과 **타뷸레이션(Tabulation)** 의 **장단점**을 각각 비교 설명하세요.

 ### 정답 및 해설

 **정답:**



 **메모이제이션 (Memoization, Top-Down):**

 * **장점:**

     * **직관적**: 재귀 코드 구조가 문제의 점화식(수학적 관계식)과 비슷해서 이해하기 쉬울 수 있습니다.

     * **필요한 것만 계산**: 딱 필요한 부분 문제의 답만 계산합니다.

 * **단점:**

     * **재귀 비용**: 함수를 계속 호출하는 데 약간의 시간과 메모리가 더 들 수 있습니다.

     * **스택 오버플로우**: 문제가 너무 복잡하면 재귀 호출이 너무 깊어져 에러가 날 수 있습니다.



 **타뷸레이션 (Tabulation, Bottom-Up):**

 * **장점:**

     * **빠름**: 단순 반복문이라 재귀보다 보통 더 빠릅니다.

     * **안전**: 스택 오버플로우 걱정이 없습니다.

 * **단점:**

     * **덜 직관적**: 반복문으로 로직을 짜는 것이 재귀보다 조금 더 생각해야 할 수 있습니다.

     * **가끔 불필요한 계산**: 최종 답에 필요 없는 부분 문제의 답까지 계산할 수도 있습니다.

 ## 💻 코드 실습 (DP 효율성 체험)



 * DP를 사용하면 얼마나 빨라지는지 직접 눈으로 확인해 봅시다!

 ### 실습 1: 피보나치 수열 속도 비교 (⭐️⭐️⭐️)



 피보나치 수열을 (1)단순 재귀, (2)메모이제이션 DP, (3)타뷸레이션 DP 로 각각 구현하고,

 `n` 값을 키워가면서 실행 시간을 비교해 보세요.



 **핵심 관찰 포인트**: `n`이 30 이상으로 커지면 단순 재귀는 매우 느려지거나

 타임아웃이 발생할 수 있지만, DP 방식들은 여전히 빠르게 계산되는 것을 확인합니다.

 이는 DP가 중복 계산을 얼마나 효과적으로 줄여주는지를 보여줍니다.

### 코드예시

````python
# %%
# 실습 1: 피보나치 수열 (DP 최적화 비교) - 코드 실행 및 관찰
import time

# 1. 단순 재귀 방식 (매우 비효율적!)
def fib_recursive(n):
    if n <= 1: return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

# 2. 메모이제이션 방식 (Top-Down DP)
memo_cache_fib = {}
def fib_memoization(n):
    if n <= 1: return n
    # 캐시 확인 (핵심!)
    if n in memo_cache_fib: return memo_cache_fib[n]
    # 계산 및 캐시 저장 (핵심!)
    memo_cache_fib[n] = fib_memoization(n - 1) + fib_memoization(n - 2)
    return memo_cache_fib[n]

# 3. 타뷸레이션 방식 (Bottom-Up DP)
def fib_tabulation(n):
    if n <= 1: return n
    # 테이블(리스트) 생성
    dp = [0] * (n + 1)
    dp[1] = 1 # 기초 값 설정
    # 테이블 채우기 (핵심!)
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

# --- 테스트 ---
n_test = 30 # 이 값을 10, 20, 30, 35 등으로 바꿔가며 실행해보세요!
print(f"--- 피보나치({n_test}) 계산 시간 비교 ---")

# 단순 재귀 (주의: n=35 이상이면 매우 오래 걸릴 수 있음)
print("1. 단순 재귀 실행 중...")
start_rec = time.time()
try:
    if n_test > 35: raise TimeoutError("n=35 초과 시 너무 오래 걸림") # 시간 제한 예시
    result_rec = fib_recursive(n_test)
    end_rec = time.time()
    print(f"   결과: {result_rec}, 시간: {end_rec - start_rec:.6f} 초")
except Exception as e:
    end_rec = time.time()
    print(f"   결과: 계산 시간 초과 또는 오류 ({e}), 경과 시간: {end_rec - start_rec:.6f} 초")

# 메모이제이션
print("\n2. 메모이제이션 실행 중...")
memo_cache_fib = {} # 캐시 초기화 중요!
start_memo = time.time()
result_memo = fib_memoization(n_test)
end_memo = time.time()
print(f"   결과: {result_memo}, 시간: {end_memo - start_memo:.6f} 초")

# 타뷸레이션
print("\n3. 타뷸레이션 실행 중...")
start_tab = time.time()
result_tab = fib_tabulation(n_test)
end_tab = time.time()
print(f"   결과: {result_tab}, 시간: {end_tab - start_tab:.6f} 초")

print("\n=> DP 방식(메모이제이션, 타뷸레이션)이 단순 재귀보다 훨씬 빠른 것을 확인!")

````

 ## 🌲🔙 백트래킹 (Backtracking)



 * **핵심 아이디어**: "모든 길을 가보긴 해야 하는데... 가망 없는 길은 미리 포기하고 돌아가자!"

 * **동작 방식 (DFS와 비슷!)**:

     1.  **일단 가보기**: 한 방향으로 쭉 가봅니다 (주로 **재귀** 호출 사용, DFS 방식).

     2.  **길 확인**: 현재 길이 최종 해답으로 이어질 **가능성**이 있는지(유망한지) 확인합니다.

     3.  **가망 없다면? (가지치기 Pruning!)**: 더 이상 가봤자 답이 안 나올 것 같으면, **포기하고 이전 갈림길로 되돌아갑니다(Backtrack)** (재귀 함수 종료). 이게 백트래킹 효율의 핵심! ✂️✨

     4.  **다른 길 가보기**: 돌아온 갈림길에서 아직 안 가본 다른 길로 다시 탐색을 시작합니다.

 * **활용 분야**: 모든 조합/순열 찾기, 길 찾기(미로), 스도쿠 풀이 등 **모든 경우의 수를 체계적으로 탐색**해야 하는 문제에 사용됩니다.

 ### 백트래킹 예시: 부분집합 만들기



 * **문제**: 주어진 집합 `[1, 2, 3]`의 모든 부분집합을 찾아보세요.

 * **백트래킹 아이디어**: 각 숫자에 대해 "이번 부분집합에 **포함시킨다**" 와 "포함시키지 않는다" 두 가지 선택을 **모두 탐색**해봅니다!

````python
# %%
# find_subsets_simple 함수는 이전 6교시 코드와 동일하므로 여기서는 설명과 함께 다시 제시합니다.

def find_subsets_simple(nums):
    """리스트의 모든 부분집합 찾기 (백트래킹)"""
    result = []         # 최종 결과 저장 리스트
    current_subset = [] # 현재 만들고 있는 부분집합

    def backtrack(start_index):
        # 1. 현재까지 만든 부분집합(current_subset)을 결과(result)에 추가
        #    -> 어떤 선택을 하든 그 상태 자체는 하나의 부분집합임!
        #    ★ 중요: .copy() 로 복사본을 넣어줘야 함
        result.append(current_subset.copy())

        # 2. 다음 선택지들을 시도 (start_index 부터 끝까지)
        for i in range(start_index, len(nums)):
            # 선택 O: 현재 숫자(nums[i])를 부분집합에 넣는다
            current_subset.append(nums[i])
            # 다음 결정을 위해 재귀 호출 (i+1 부터 시작해서 중복 선택 방지)
            backtrack(i + 1)
            # 선택 X (백트래킹): 재귀 호출이 끝나고 돌아오면, 넣었던 숫자를 다시 뺀다
            #                    -> 그래야 다음 for 루프에서 다른 선택을 시도할 수 있음!
            current_subset.pop()

    print("--- 부분집합 생성 시작 ---")
    backtrack(0) # 0번 인덱스부터 탐색 시작
    print("--- 부분집합 생성 종료 ---")
    return result

# 테스트
input_nums = [1, 2, 3]
all_subsets = find_subsets_simple(input_nums)
print(f"\n{input_nums}의 모든 부분집합:")
print(all_subsets) # [[], [1], [1, 2], [1, 2, 3], [1, 3], [2], [2, 3], [3]] (순서 다를 수 있음)
````

 ## ❓ 객관식, 단답형 및 서술형 문제



 * (백트래킹 관련 문제만 남깁니다.)

 ### 🧐 객관식 문제 1 (난이도: ⭐️⭐️⭐️)



 **백트래킹(Backtracking)** 알고리즘에 대한 설명으로 **가장 적절하지 않은** 것은 무엇인가요?



 1.  해를 찾는 과정에서 가능성이 없는 경로는 포기하고 이전 단계로 되돌아간다.

 2.  주로 깊이 우선 탐색(DFS) 방식으로 상태 공간 트리를 탐색한다.

 3.  가지치기(Pruning)를 통해 탐색의 효율성을 높일 수 있다.

 4.  항상 최적의 해를 찾는 것을 보장하는 최적화 기법이다.

 5.  조합, 순열 생성, 제약 만족 문제 등에 활용될 수 있다.

 ### 정답 및 해설

 **정답: 4번**



 **해설:**

 백트래킹은 해가 될 가능성이 있는 모든 후보 경로를 탐색하는 **탐색 기법**이지, 반드시 **최적의 해**를 찾는 것을 보장하는 **최적화 기법**은 아닙니다. 백트래킹은 모든 가능한 해를 찾거나, 조건을 만족하는 첫 번째 해를 찾는 데 주로 사용됩니다. 최적화 문제에는 동적 계획법(DP)이나 탐욕 알고리즘 등이 더 적합한 경우가 많습니다.

 ### ✍️ 단답형 / 서술형 문제 1 (난이도: ⭐️⭐️⭐️⭐️)



 **백트래킹(Backtracking)** 에서 **가지치기(Pruning)** 가 무엇인지 설명하고, 가지치기가 왜 백트래킹 알고리즘의 **효율성 향상**에 중요한지 설명하세요.

 ### 정답 및 해설

 **정답:**



 **가지치기 (Pruning)란:**

 백트래킹으로 여러 가능성을 탐색하다가, "어? 이쪽 길로 계속 가봤자 어차피 답이 안 나오겠는데?" 라고 **미리 판단**될 때, **더 이상 그 길로는 가지 않고 되돌아 나오는 것**을 말합니다. 쓸모없는 나뭇가지를 잘라내는 것과 비슷합니다.



 **효율성 향상에 중요한 이유:**

 백트래킹은 기본적으로 모든 가능성을 다 시도해보는 방식이라 경우의 수가 많으면 엄청나게 느려질 수 있습니다. 가지치기는 이처럼 **답이 될 가능성이 없는 길을 미리 차단**해주기 때문에, **불필요한 탐색 시간을 대폭 줄여줍니다.** 즉, 탐색해야 할 범위를 효과적으로 좁혀서 훨씬 **빠르게 답을 찾거나 모든 답을 탐색**할 수 있도록 도와주는 핵심 기술입니다.

 ### 실습 1: 부분집합 생성 코드 이해 및 디버깅 (⭐️⭐️⭐️)



 위에서 제공된 `find_subsets_simple` 함수의 동작 과정을 주석 또는 `print` 문을 추가하여 단계별로 따라가 보세요.



 **해볼 것:**

 1.  `backtrack` 함수가 호출될 때 `start_index`와 `current_subset`이 어떻게 변하는지 관찰합니다.

 2.  `result.append(current_subset.copy())` 가 왜 필요한지 생각해 봅니다 (`.copy()` 없이 실행하면 어떻게 될까요?).

 3.  `for` 루프 안에서 `append` -> `backtrack` 호출 -> `pop` 과정이 어떻게 "선택"과 "선택 취소(백트래킹)"를 구현하는지 이해합니다.

 4.  만약 `input_nums = [1, 2]` 라면 `result`는 어떻게 채워질까요? 직접 손으로 따라가 보거나 코드를 수정하여 실행해 보세요.

### 코드예시

````python
# %%
# 실습 1: 부분집합 생성 코드 이해 및 디버깅
def find_subsets_debug(nums):
    result = []
    current_subset = []
    n = len(nums)

    def backtrack(start_index):
        # 현재 상태 출력 (디버깅용)
        print(f"backtrack(start_index={start_index}), current_subset={current_subset}")

        # 현재 부분집합을 결과에 추가
        result.append(current_subset.copy())
        print(f"  -> result에 추가: {current_subset}")

        # 다음 선택지 탐색
        print(f"  -> for 루프 시작 (i = {start_index}부터 {n-1}까지)")
        for i in range(start_index, n):
            # 선택
            num_to_add = nums[i]
            current_subset.append(num_to_add)
            print(f"    -> {num_to_add} 추가: {current_subset}. 다음 호출 backtrack({i + 1})")

            # 재귀 호출
            backtrack(i + 1)

            # 선택 취소 (백트래킹)
            removed = current_subset.pop()
            print(f"    <- {removed} 제거 (백트래킹 후): {current_subset}. (i={i} 루프 다음으로)")
        print(f"  -> for 루프 종료 (start_index={start_index})")


    print("--- 부분집합 생성 시작 (디버그 모드) ---")
    backtrack(0)
    print("--- 부분집합 생성 종료 ---")
    return result

# 테스트 (작은 입력으로)
input_nums_debug = [1, 2]
print(f"\n입력: {input_nums_debug}")
all_subsets_debug = find_subsets_debug(input_nums_debug)
print(f"\n{input_nums_debug}의 모든 부분집합 (최종 결과):")
print(all_subsets_debug) # 예상: [[], [1], [1, 2], [2]]
````

 ## 💻 종합 실습 문제



 ### 문제 1: 미로 탈출 최단 경로 찾기 (BFS 활용) (난이도: ⭐️⭐️⭐️⭐️)



 **상황:** N x M 크기의 미로(`maze`, 0:길, 1:벽)에서 출발점 `start` 부터 도착점 `end` 까지 이동 가능한 **최단 경로의 길이(이동 칸 수)** 를 구하세요. 상하좌우 이동 가능. 도달 불가 시 -1 반환.



 **데이터 예시:**

 ```python

 maze = [

     [0, 1, 0, 0, 0],

     [0, 0, 0, 1, 0],

     [1, 1, 0, 1, 0],

     [0, 0, 0, 0, 0]

 ]

 start = (0, 0)

 end = (3, 4)

 ```

 **예상 출력:** `8`



 **🤔 필요 기술:** BFS, 큐(deque), 방문 처리(2D 배열 or Set)

### 코드예시

````python
# %%
# 문제 1 코드 구현 영역 (이전 8교시 실습 1 코드 참고 또는 직접 작성)
from collections import deque

def solve_maze_shortest_path(maze, start, end):
    # 함수 내용을 작성하세요 (이전 8교시 코드를 참고하거나 직접 구현)
    if not maze: return -1
    R, C = len(maze), len(maze[0])
    start_r, start_c = start
    end_r, end_c = end

    if maze[start_r][start_c] == 1 or maze[end_r][end_c] == 1:
        return -1

    visited = [[False] * C for _ in range(R)]
    queue = deque([(start_r, start_c, 0)])
    visited[start_r][start_c] = True

    dr = [-1, 1, 0, 0]
    dc = [0, 0, -1, 1]

    while queue:
        r, c, dist = queue.popleft()
        if r == end_r and c == end_c:
            return dist

        for i in range(4):
            nr, nc = r + dr[i], c + dc[i]
            if 0 <= nr < R and 0 <= nc < C and \
               maze[nr][nc] == 0 and not visited[nr][nc]:
                visited[nr][nc] = True
                queue.append((nr, nc, dist + 1))

    return -1

# 테스트
maze_data = [
    [0, 1, 0, 0, 0],
    [0, 0, 0, 1, 0],
    [1, 1, 0, 1, 0],
    [0, 0, 0, 0, 0]
]
start_pos = (0, 0)
end_pos = (3, 4)
print(f"--- 문제 1: 미로 찾기 ---")
shortest_dist = solve_maze_shortest_path(maze_data, start_pos, end_pos)
print(f"최단 경로 길이: {shortest_dist}")

````

 ### 문제 2: 연결된 그룹(컴포넌트) 개수 찾기 (BFS 또는 DFS 활용) (난이도: ⭐️⭐️⭐️⭐️)



 **상황:** 친구 관계(무방향) 정보가 주어졌을 때, 서로 연결된 친구 그룹(Connected Components)이 총 몇 개인지 찾는 함수를 작성하세요. 예를 들어, A-B, B-C, D-E 가 주어지면 {A, B, C} 그룹과 {D, E} 그룹으로 총 2개의 그룹이 있습니다.

 **데이터 예시:**

 all_friends = {'A', 'B', 'C', 'D', 'E', 'F', 'G'}

 **🤔 필요 기술:** 그래프 표현(인접 리스트), BFS 또는 DFS, 방문 처리(Set)

### 코드 예시

````python
# %%
# 문제 2 코드 구현 영역
from collections import defaultdict, deque

def count_friend_groups(all_friends, relationships):
    """친구 그룹(Connected Components)의 개수를 세는 함수"""
    if not all_friends:
        return 0

    # 1. 인접 리스트 그래프 생성
    graph = defaultdict(list)
    for u, v in relationships:
        graph[u].append(v)
        graph[v].append(u)

    visited = set() # 방문한 친구 기록
    group_count = 0 # 그룹 개수

    # 2. 모든 친구에 대해 확인
    for friend in all_friends:
        # 3. 아직 방문하지 않은 친구 발견 시
        if friend not in visited:
            group_count += 1 # 새로운 그룹 발견!
            print(f"\n새 그룹 발견! 시작: {friend}")
            # 해당 친구부터 시작하여 연결된 모든 친구 방문 (BFS 또는 DFS)
            # 여기서는 BFS 사용 예시
            queue = deque([friend])
            visited.add(friend)
            print(f"  방문 시작: {friend}")
            while queue:
                current_friend = queue.popleft()
                for neighbor in graph[current_friend]:
                    if neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
                        print(f"  방문 추가: {neighbor}")

    return group_count

# 테스트
friends = {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'} # H 추가 (혼자 있는 친구)
rels = [('A', 'B'), ('B', 'C'), ('D', 'E'), ('F', 'G')]

print(f"\n--- 문제 2: 친구 그룹 수 찾기 ---")
num_groups = count_friend_groups(friends, rels)
print(f"\n총 친구 그룹 수: {num_groups}") # 예상: 4 ({A,B,C}, {D,E}, {F,G}, {H})
````

 ### 문제 3: 애너그램 그룹핑 (딕셔너리, 정렬 활용) (난이도: ⭐️⭐️⭐️)



 **상황:** 문자열 리스트 `strs` 가 주어졌을 때, 애너그램(Anagram, 문자를 재배열하여 만들 수 있는 단어) 관계인 단어들끼리 그룹핑하는 함수를 작성하세요.



 **요구사항:**

 1. 문자열 리스트 `strs`를 입력받습니다.

 2. 애너그램끼리 묶인 리스트들의 리스트를 반환합니다. (그룹 내 순서, 그룹 간 순서는 무관)



 **데이터 예시:**

 ```python

 strs = ["eat", "tea", "tan", "ate", "nat", "bat"]

 ```

 **예상 출력:** `[['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]` (순서는 다를 수 있음)



 **🤔 어떤 자료구조/알고리즘을 사용하는 것이 좋을까요?**

 * 애너그램은 **구성 문자를 정렬**하면 모두 같은 문자열이 됨! → **정렬 (sorted)**

 * 정렬된 문자열을 **키(Key)** 로 사용하고, 원본 단어들을 **값(Value) 리스트**로 저장 → **딕셔너리 (defaultdict)**

### 코드 예시

````python
# %%
# 문제 3 코드 구현 영역
from collections import defaultdict

def group_anagrams(strs: list[str]) -> list[list[str]]:
    """애너그램 단위로 단어들을 그룹핑하는 함수"""
    anagram_groups = defaultdict(list) # Key: 정렬된 문자열, Value: 원본 단어 리스트

    # 1. 각 단어(s)를 순회
    for s in strs:
        # 2. 단어의 문자를 정렬하여 키 생성 (''.join(sorted(s)))
        sorted_chars = sorted(s) # ['a', 'e', 't']
        key = "".join(sorted_chars) # "aet"
        # print(f"단어: {s}, 정렬된 키: {key}") # 확인용

        # 3. 정렬된 키를 기준으로 딕셔너리에 원본 단어 추가
        anagram_groups[key].append(s)

    # 4. 딕셔너리의 값들(그룹 리스트)만 추출하여 최종 결과 반환
    return list(anagram_groups.values())

# 테스트
str_list = ["eat", "tea", "tan", "ate", "nat", "bat", "cat", "tac"]
print(f"\n--- 문제 3: 애너그램 그룹핑 ---")
grouped = group_anagrams(str_list)
print(f"입력: {str_list}")
print(f"결과: {grouped}") # [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat'], ['cat', 'tac']] (순서 다를 수 있음)

````
