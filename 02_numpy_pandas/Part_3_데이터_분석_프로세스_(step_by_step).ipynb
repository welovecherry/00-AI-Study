````python
# ‚ñ∂ Target label ÏÉùÏÑ± Sample Code (1-loop)
i = '2011-01'

# ‚ñ∂ Íµ¨ÌïòÍ≥†Ïûê ÌïòÎäî Í∏∞Ï§ÄÏõî Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
df_left = df[df['bsym'] == i]

# ‚ñ∂ Í∏∞Ï§ÄÎÖÑÏõî +1m
bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')

# ‚ñ∂ Í∏∞Ï§ÄÎÖÑÏõî +1m, CustomerIDÎäî 1ÏõîÏùò Íµ¨Îß§Ìïú Í≥†Í∞ù
df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
df_right['target'] = 'Y'
df_right.columns = ['CustomerID', 'target']

# ‚ñ∂ Í∏∞Ï§ÄÎÖÑÏõî DataÏóê Í∏∞Ï§ÄÎÖÑÏõî +1M Í≥†Í∞ùÏùÑ JoinÌïòÎ©¥ Target Data ÏÉùÏÑ± ÏôÑÎ£å
df_merge = pd.merge(df_left, df_right, how='left')
df_merge['target'] = df_merge['target'].fillna('N')
````

````python
# ‚ñ∂ forÎ¨∏ ÌôúÏö© Ï†ÑÏ≤¥ Data Set ÎåÄÏÉÅ Target label ÏÉùÏÑ±
loop_list = list(df['bsym'].unique())
df_all = []

for i in loop_list :
  df_left = df[df['bsym'] == i]
  bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')
  df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
  df_right['target'] = 1
  df_right.columns = ['CustomerID', 'target']
  df_merge = pd.merge(df_left, df_right, how='left')
  df_merge['target'] = df_merge['target'].fillna(0)

  df_all.append(df_merge)

df_all = pd.concat(df_all)
````

```
* Î™®Îç∏ÎßÅÏùò Î™©Ï†ÅÏù∏ ÏÑ±Îä•Ìñ•ÏÉÅÏùÑ ÏúÑÌï¥ FeatureÎ•º ÏÉùÏÑ±, ÏÑ†ÌÉù, Í∞ÄÍ≥µÌïòÎäî ÏùºÎ†®Ïùò Î™®Îì† ÌôúÎèô

> **Target Î≥ÄÏàòÏôÄÏùò ÏùòÎØ∏ÏûàÎäî Î≥ÄÏàò ÏÑ†ÌÉùÌïòÎäî Î∞©Î≤ï**

```
* Regression(ÌöåÍ∑Ä) : ÏÉÅÍ¥ÄÍ≥ÑÏàò Î∂ÑÏÑùÏùÑ ÌÜµÌï¥ Ïú†ÏùòÎØ∏Ìïú Î≥ÄÏàò ÏÑ†ÌÉù

> **IV (Information Value)**

```
* Feature ÌïòÎÇòÍ∞Ä Good(Target)Í≥º Bad(Non-target)ÏùÑ Ïûò Íµ¨Î∂ÑÌï¥Ï§Ñ Ïàò ÏûàÎäîÏßÄÏóê ÎåÄÌïú Ï†ïÎ≥¥ÎüâÏùÑ ÌëúÌòÑÌïòÎäî Î∞©Î≤ïÎ°†
* IV ÏàòÏπòÍ∞Ä ÌÅ¥ÏàòÎ°ù TargetÍ≥º Non-targetÏùÑ Ïûò Íµ¨Î∂ÑÌï† Ïàò ÏûàÎäî Ï†ïÎ≥¥ÎüâÏù¥ ÎßéÏùÄ FeatureÏù¥Í≥†, IV ÏàòÏπòÍ∞Ä ÏûëÏùÑÏàòÎ°ù Ï†ïÎ≥¥ÎüâÏù¥ Ï†ÅÏùÄ Feature
* (Target data Íµ¨ÏÑ±ÎπÑ - Non-target data Íµ¨ÏÑ±ÎπÑ) * WoE
* WoE(Weight of Evidence) = ln(Target data Íµ¨ÏÑ±ÎπÑ / Non-target data Íµ¨ÏÑ±ÎπÑ)
* ln(ÏûêÏó∞Î°úÍ∑∏)Î•º Ï∑®ÌïòÎäî Ïù¥Ïú†? : WoEÏùò ÏµúÎåÄ, ÏµúÏÜåÍ∞íÏùò Î≤îÏúÑÎ•º ÎßûÏ∂∞Ï£ºÍ∏∞ ÏúÑÌï¥ (1~‚àû , -‚àû~1)
```

> ### **‚îî Classification Î¨∏Ï†ú (IV)**

````python
# ‚ñ∂ Binning
total_amt_quantile  = df_mart['total_amt'].quantile(q=list(np.arange(0.05,1,0.05)), interpolation = 'nearest')
# total_amt_quantile

# ‚ñ∂ Í∑∏Î£π Ï¥àÍ∏∞Ìôî
df_mart['total_amt_grp'] = 0

# ‚ñ∂ 40%, 75% Í∏∞Ï§ÄÏúºÎ°ú Ï¥ù 3Í∞ú Íµ¨Í∞ÑÌôî ÏßÑÌñâ
for i in  range(len(df_mart['total_amt'])) :
  if df_mart.loc[i, "total_amt"] <= total_amt_quantile.iloc[7] : # 40%
    df_mart.loc[i, 'total_amt_grp'] = 1
  elif df_mart.loc[i, "total_amt"] <= total_amt_quantile.iloc[14] : # 75%
    df_mart.loc[i, 'total_amt_grp'] = 2
  else : df_mart.loc[i, 'total_amt_grp'] = 3
````

````python
df_mart.groupby('total_amt_grp')['target'].agg(['sum', 'count'])
````

````python
# ‚ñ∂ non-target labeling
df_mart['n_target'] = np.where(df_mart['target']==0, 1, 0)

# ‚ñ∂ target , non-target count
iv = df_mart.groupby('total_amt_grp')[['target', 'n_target']].sum().reset_index()

# ‚ñ∂ good_pct / bad_pct Í≥ÑÏÇ∞
iv['good_pct'] = iv['target'] / iv['target'].sum()
iv['bad_pct'] = iv['n_target'] / iv['n_target'].sum()

# ‚ñ∂ t_ratio
iv['t_ratio'] = iv['target'] / (iv['target'] + iv['n_target'])

iv
````

````python
# ‚ñ∂ GroupÎ≥Ñ IV Í≥ÑÏÇ∞
import math

for i in range(len(iv['total_amt_grp'])) :
  iv.loc[i, 'iv'] = (iv.loc[i,'good_pct'] - iv.loc[i,'bad_pct']) * math.log(iv.loc[i,'good_pct']/iv.loc[i,'bad_pct'])

iv
````

````python
# forÎ¨∏ ÌôúÏö© ÏûêÎèôÌôî
integer_list = ['total_amt', 'max_amt', 'min_amt', 'total_cnt','max_cnt', 'min_cnt', 'total_qty', 'max_qty', 'min_qty']
iv_df = []

for val in integer_list :
    # ‚ñ∂ Binning
    quantile  = df_mart[val].quantile(q=list(np.arange(0.05,1,0.05)), interpolation = 'nearest')

    # ‚ñ∂ Í∑∏Î£π Ï¥àÍ∏∞Ìôî
    df_mart['grp'] = 0

    # ‚ñ∂ 40%, 75% Í∏∞Ï§ÄÏúºÎ°ú Ï¥ù 3Í∞ú Íµ¨Í∞ÑÌôî ÏßÑÌñâ
    for i in  range(len(df_mart[val])) :
      if df_mart.loc[i, val] <= quantile.iloc[7] :
        df_mart.loc[i, 'grp'] = 1
      elif df_mart.loc[i, val] <= quantile.iloc[14] :
        df_mart.loc[i, 'grp'] = 2
      else : df_mart.loc[i, 'grp'] = 3

    # ‚ñ∂ non target labeling
    df_mart['n_target'] = np.where(df_mart['target']==0, 1, 0)

    # ‚ñ∂ target , non-target count
    iv = df_mart.groupby('grp')[['target', 'n_target']].sum().reset_index()

    # ‚ñ∂ good_pct / bad_pct Í≥ÑÏÇ∞
    iv['good_pct'] = iv['target'] / iv['target'].sum()
    iv['bad_pct'] = iv['n_target'] / iv['n_target'].sum()

    # ‚ñ∂ t_ratio
    iv['t_ratio'] = iv['target'] / (iv['target'] + iv['n_target'])

    iv['val'] = val

    for i in range(len(iv['grp'])) :
      iv.loc[i, 'iv'] = (iv.loc[i,'good_pct'] - iv.loc[i,'bad_pct']) * math.log(iv.loc[i,'good_pct']/iv.loc[i,'bad_pct'])

    iv_df.append(iv)

iv_df=pd.concat(iv_df)
iv_df
````

````python
# ‚ñ∂ FeatureÎ≥Ñ Ï†ÑÏ≤¥ IV Í∞íÏùÑ sumÌï¥ÏÑú ÎπÑÍµê
iv_df.groupby('val')['iv'].sum().sort_values(ascending=False)
````

````python
# ‚ñ∂ ÏµúÏ¢Ö Ï∂îÏ∂úÎêú Mart read
import pandas as pd
df_mart = pd.read_csv('df_mart.csv')
df_mart.head()
````

````python
# ‚ñ∂ ÌïôÏäµÏóê ÌïÑÏöîÏóÜÎäî Column Ï†úÍ±∞
df_mart = df_mart.drop(['bsym', 'CustomerID', 'total_amt_grp', 'n_target', 'grp'], axis=1)
df_mart.head()
````

````python
# ‚ñ∂ Î™®Îç∏ÎßÅÏùÑ ÌïôÏäµÌïòÍ∏∞ ÏúÑÌïú Feature(X)ÏôÄ Target(Y)Îç∞Ïù¥ÌÑ∞Î•º Íµ¨Î∂ÑÌïòÎäî Îã®Í≥Ñ
from sklearn.model_selection import train_test_split

X = df_mart.drop(['target'], axis=1)
Y = df_mart['target']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=1234)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
````

````python
# ‚ñ∂ Ïà´ÏûêÌòï(Integer), Î≤îÏ£ºÌòï(Categorical) Î≥ÄÏàò Î∂ÑÌï†
numerical_list=[]
categorical_list=[]

for i in df_mart.columns :
  if df_mart[i].dtypes == 'O' :
    categorical_list.append(i)
  else :
    numerical_list.append(i)

print("categorical_list :", categorical_list)
print("numerical_list :", numerical_list)
````

````python
# ‚ñ∂ (1) One hot encoding

from sklearn.preprocessing import OneHotEncoder

for col in categorical_list :
  encoder = OneHotEncoder()
  encoder.fit(x_train[[col]])
  onehot_train = pd.DataFrame(encoder.transform(x_train[[col]]).toarray(), columns = encoder.get_feature_names(), index=x_train.index)
  onehot_test = pd.DataFrame(encoder.transform(x_test[[col]]).toarray(), columns = encoder.get_feature_names(), index=x_test.index)
  # Í∏∞Ï°¥ ColÏùÄ ÏÇ≠Ï†ú
  x_train = pd.concat([x_train,onehot_train], axis = 1).drop(columns = [col])
  x_test = pd.concat([x_test,onehot_test], axis = 1).drop(columns = [col])
````

````python
x_train.iloc[0,:], x_test.iloc[0,:]
````

````python
# ‚ñ∂ (2) label encoding

from sklearn.preprocessing import LabelEncoder

for col in categorical_list :
  encoder = LabelEncoder()
  encoder.fit(x_train[[col]])

  # Í∏∞Ï°¥ Col ÎåÄÏ≤¥
  x_train[col] = encoder.transform(x_train[col])
  x_test[col] = encoder.transform(x_test[col])
````

````python
# ‚ñ∂ BayesianOptimization
import numpy as np
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score


def model_evaluate(n_estimators, maxDepth):
    clf = RandomForestClassifier(
        n_estimators= int(n_estimators),
        max_depth= int(maxDepth))
    scores = cross_val_score(clf, x_train, y_train, cv=5, scoring='roc_auc')
    return np.mean(scores)


def bayesOpt(x_train, y_train):
    clfBO = BayesianOptimization(model_evaluate, {'n_estimators':  (100, 200),
                                                  'maxDepth': (2, 4)
                                                 })
    clfBO.maximize(init_points=5, n_iter=10)
    print(clfBO.res)

bayesOpt(x_train, y_train)
````
