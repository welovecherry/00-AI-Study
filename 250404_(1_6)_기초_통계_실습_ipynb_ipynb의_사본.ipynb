```python
import pandas as pd

df = pd.read_csv('gapminder.tsv', sep = '\t')
df
```

```python
# 연도(year)를 기준으로 그룹화하고 인구(pop) 열의 도수분포표 생성
print(df.describe())
frequency_table = df.groupby('year')['pop'].sum().reset_index()

# 결과 출력
print("\nfrequency_table")
print(frequency_table)
```
