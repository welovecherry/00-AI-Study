### 3-2 ì›-í•« ì¸ì½”ë”© (One-Hot Encoding)

> ë²”ì£¼í˜• ë³€ìˆ˜ë¥¼ 0ê³¼ 1ë¡œë§Œ êµ¬ì„±ëœ ì´ì§„(Binary) í˜•íƒœë¡œ ë³€í™˜í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. ë³€ìˆ˜ì˜ ê³ ìœ (Unique) ë²”ì£¼ í¬ê¸°ì™€ ë™ì¼í•œ ì´ì§„ ë²¡í„°ê°€ ìƒì„±ë©ë‹ˆë‹¤.<br> ë²”ì£¼ì— í•´ë‹¹í•˜ëŠ” ê°’ì€ 1, ë‚˜ë¨¸ì§€ëŠ” 0ìœ¼ë¡œ ë³€í™˜í•˜ê²Œ ë©ë‹ˆë‹¤.

- ì¦‰, ì•Œê³ ë¦¬ì¦˜ ëª¨ë¸ì´ ë³€ìˆ˜ì˜ ì˜ë¯¸ë¥¼ ì •í™•í•˜ê²Œ íŒŒì•…í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

- ê³ ìœ  ë²”ì£¼ì˜ í¬ê¸°ê°€ ëŠ˜ì–´ë‚  ë•Œë§ˆë‹¤ ë²¡í„°ì˜ ì°¨ì›ì´ ëŠ˜ì–´ë‚˜ëŠ” ë¬¸ì œê°€ ì¡´ì¬í•©ë‹ˆë‹¤.

````python
onehot_data = OHLCV_data.copy()

# scikit-learnì—ì„œ ì œê³µí•˜ëŠ” OneHotEncoder í´ë˜ìŠ¤ë¥¼ í™œìš©í•´ ì¢…ëª©(code)ì—´ì˜ One-Hot Encodingì„ ì§„í–‰í•´ ë´…ì‹œë‹¤.
# OneHotEncoderì˜ í´ë˜ìŠ¤ ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
encoder = OneHotEncoder()

# ì›-í•« ì¸ì½”ë”©ì—ì„œëŠ” 2ì°¨ì› í˜•íƒœì˜ ë°°ì—´ë¡œ ì…ë ¥í•´ì•¼ í•˜ê¸° ë•Œë¬¸ì— ëŒ€ê´„í˜¸([])ê°€ 2ê°œê°€ ë“¤ì–´ê°‘ë‹ˆë‹¤.
onehot = encoder.fit_transform(onehot_data[["code"]])

# ë°°ì—´í˜•íƒœë¡œ ì „í™˜
onehot_array = onehot.toarray()

# ì¢…ëª©ì˜ ê³ ìœ  ë²”ì£¼ìˆ˜ 999ê°œì¸ ê²ƒê³¼ ë™ì¼í•˜ê²Œ 999ì°¨ì›ì˜ ì›-í•« ì¸ì½”ë”© ë°°ì—´ì´ ìƒê²¼ìŠµë‹ˆë‹¤.
# ì´ëŠ” ë°ì´í„°ì…‹ì˜ ê·œëª¨ê°€ ì»¤ì§ˆ ë•Œ ë©”ëª¨ë¦¬ì— ì˜í–¥ì„ ì¤„ìˆ˜ ìˆìŠµë‹ˆë‹¤.
display(onehot_array)
display(f"ì¢…ëª© ê³ ìœ  ë²”ì£¼ ìˆ˜ : {len(onehot_data['code'].unique())}")
display(f"ì›-í•« ì¸ì½”ë”©ì˜ ì°¨ì› ìˆ˜ : {onehot_array.shape}")
````

Count Encodingì´ë¼ê³ ë„ ë¶ˆë¦½ë‹ˆë‹¤.

#### ğŸ“ ì„¤ëª… : (3-4)
- ë¹ˆë„ë¼ëŠ” ìˆ˜ì¹˜ì ì¸ ì˜ë¯¸ë¥¼ ë³€ìˆ˜ì— ë¶€ì—¬ ê°€ëŠ¥í•©ë‹ˆë‹¤.
- í•œê°€ì§€ ë³€ìˆ˜ë¡œ í‘œí˜„ ê°€ëŠ¥í•˜ê¸° ë•Œë¬¸ì— ë©”ëª¨ë¦¬ ê´€ë¦¬ ì¸¡ë©´ì—ì„œ íš¨ìœ¨ì ì…ë‹ˆë‹¤.

---

- ë‹¤ë¥¸ íŠ¹ì„±ì˜ ì˜ë¯¸ë¥¼ ì§€ë‹ˆê³  ìˆì–´ë„ ë¹ˆë„ê°€ ê°™ìœ¼ë©´, ë‹¤ë¥¸ ë²”ì£¼ê°„ ì˜ë¯¸ê°€ ë™ì¼í•˜ê²Œ ì¸ì‹ë  ìˆ˜ ìˆëŠ” ê°€ëŠ¥ì„±ì´ ìˆìŠµë‹ˆë‹¤.

- scikit-learnì—ì„œëŠ” ì•„ì‰½ê²Œë„ FrequencyEncoderë¥¼ ì œê³µí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
- ëŒ€ì‹  category_encodersë¼ëŠ” íŒ¨í‚¤ì§€ë¥¼ ì„¤ì¹˜í•˜ì—¬ CountEncoderë¥¼ ì‚¬ìš©í•´ë´…ì‹œë‹¤.

````python
frequency_data = OHLCV_data.copy()

# CountEncoderë¥¼ í´ë˜ìŠ¤ ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
# colsì—ëŠ” ì ìš©í•  ì—´ì„ ëª…ì‹œí•´ì¤ë‹ˆë‹¤.
encoder = CountEncoder(cols=['code'])

# ë³€í™˜ì‹œì¼œì¤ì‹œë‹¤.
frequency = encoder.fit_transform(frequency_data['code'])

frequency_data["FrequencyCode"] = frequency

display(frequency_data)
````

ë²”ì£¼í˜• ë³€ìˆ˜ê°€ ì—°ì†ì ì¸ íŠ¹ì„±ì„ ê°€ì§„ ê°’ìœ¼ë¡œ ë³€í™˜ë©ë‹ˆë‹¤. <br>

#### ğŸ“ ì„¤ëª… : (3-5)
- ë²”ì£¼ê°„ ìˆ˜ì¹˜ì ì¸ ì˜ë¯¸ë¥¼ ë³€ìˆ˜ì— ë¶€ì—¬ ê°€ëŠ¥í•©ë‹ˆë‹¤.
- íƒ€ê²Ÿ ë³€ìˆ˜ë¼ëŠ” ì¶”ê°€ì ì¸ ì •ë³´ë¥¼ ê°€ì§„ ë³€ìˆ˜ì— ì˜ì¡´í•˜ë¯€ë¡œ íŒŒìƒ ë³€ìˆ˜ì™€ ë¹„ìŠ·í•©ë‹ˆë‹¤.

- ê³¼ì í•©ì´ë‚˜ Data-Leakageì— ëŒ€í•œ ë¬¸ì œê°€ ë°œìƒí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

- pandasë¥¼ ì´ìš©í•´ Target Encodingì„ ì§„í–‰í•©ë‹ˆë‹¤.
- ë²”ì£¼í˜• ë³€ìˆ˜ ì™¸ì—, í‰ê· ê°’ì„ êµ¬í•˜ê¸° ìœ„í•œ ì—°ì†ì  íŠ¹ì„±ì„ ê°€ì§„ íƒ€ê²Ÿ ë³€ìˆ˜ê°€ í•„ìš”í•©ë‹ˆë‹¤.
- ì—¬ê¸°ì„œëŠ” Open(ì‹œê°€) ì—´ì„ íƒ€ê²Ÿ ë³€ìˆ˜ë¼ê³  ê°€ì •í•˜ê² ìŠµë‹ˆë‹¤.

````python
target_data = OHLCV_data.copy()

# ìš°ì„  pandasì—ì„œ ì œê³µí•˜ëŠ” groupbyë¥¼ í†µí•´ codeì—´ì„ ê¸°ì¤€ìœ¼ë¡œ Openì—´ì˜ í‰ê· ì„ êµ¬í•´ë´…ì‹œë‹¤.
target = target_data.groupby(OHLCV_data['code'])['Open'].mean()

# mapì„ ì´ìš©í•´ ë³€í™˜í•´ì¤ë‹ˆë‹¤.
target_data['TargetCode'] = target_data['code'].map(target)

display(target_data)
````
